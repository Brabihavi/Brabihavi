```python
import os
import shutil
import math
import win32com
from win32com.client import *
import ezdxf
from ezdxf.math import Vec3
from ezdxf.render import forms
from ezdxf.enums import TextEntityAlignment
from ezdxf.math import ConstructionArc
from ezdxf.math import Matrix44
from ezdxf.upright import upright
import tkinter
from tkinter import ttk
import time
from tkinter import messagebox
import subprocess
import ctypes
import time
from datetime import datetime, timedelta
import psutil
import win32con



the_dict = {}





def get_recently_started_tasks():
    c = 0    
    cmd = 'tasklist /FI "IMAGENAME eq acad.exe"'
    output = subprocess.check_output(cmd, shell=True).decode('utf-8')
    
    if 'acad.exe' in output:
        c += 1

    
    if c == 0:
        path_name = r"C:/Program Files/Autodesk/AutoCAD 2022/acad.exe"
        startupinfo = subprocess.STARTUPINFO()
        startupinfo.dwFlags |= subprocess.STARTF_USESHOWWINDOW
        startupinfo.wShowWindow = win32con.SW_SHOW

        process = subprocess.Popen([path_name], startupinfo=startupinfo)
        time.sleep(30)
    else:
        return


def check_intersection(line):
    liness = []
    for i in range(len(line)):
        ll = []
        line1 = LineString([line[i][0], line[i][1]])
        for j in range(i+1, len(line)):
            line2 = LineString([line[j][0], line[j][1]])
            if line1.intersects(line2):
                ll.append(line[i][0])
                ll.append(line[i][1])
                ll.append(line[j][0])
                ll.append(line[j][1])
        if ll:
            liness.append(ll)
    return liness
    
                
                

def max_abs(numbers):
    """
    Returns the maximum value among a list of numbers.
    """
    max_val = None
    for num in numbers:
        if max_val is None or num > max_val:
            max_val = num
    return max_val

def min_abs(numbers):
    """
    Returns the minimum value among a list of numbers.
    """
    min_val = None
    for num in numbers:
        if min_val is None or num < min_val:
            min_val = num 
    return min_val




    

def line_length(x,y):
    total_distance = 0.0
    for i in range(len(x)-1):
        x1 = x[i]
        x2 = x[i+1]
        y1 = y[i]
        y2 = y[i+1]
        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        total_distance += distance
        if i == len(x)-2:
            distance = math.sqrt((x[i+1] - x[0])**2 + (y[i+1] - y[0])**2)
            total_distance += distance
        else:
            pass
    return total_distance 

def print_tree_reviw ():
    for i in tree.get_children():
        tree.delete(i)

    for name, number in the_dict.items():
        tree.insert("", "end", text=name, values=(number))
    



def update_the_dict(name,number):
    the_dict[name] = number+1
    print_tree_reviw()
    
    

def update_i_distance_cnc(x1 , x2 , x4 , x5):
    distance_c = x1 + 8 
    i_value = x2 + 1
    distance_x_c = x5
    with open('C:/Users/A.Mohammadlou/Desktop/software_test/previous_values.txt', 'r') as f:
        contents = f.read()
        numbers_str = contents.split('\n')
        distance_y = float(numbers_str[1])
        distance_x = float(numbers_str[3])
    f.close()
    with open('C:/Users/A.Mohammadlou/Desktop/software_test/previous_values.txt', 'w') as f:
        f.write(str(i_value))
        f.write('\n')
        f.write(str(distance_y))
        f.write('\n')
        f.write(str(distance_c))
        f.write('\n')
        f.write(str(distance_x))
        f.write('\n')
        f.write(str(distance_x_c))
        
    update_the_dict(x4,x2)
    
    
def update_i_distance_nasb(x1 , x2, x3 , x4 , x5):
    distance_y = x1 + 8 + x3 
    i_value = x2 + 1
    distance_x = x5
    with open('C:/Users/A.Mohammadlou/Desktop/software_test/previous_values.txt', 'r') as f:
        contents = f.read()
        numbers_str = contents.split('\n')
        distance_c = float(numbers_str[2])
        distance_x_c = float(numbers_str[4])
    f.close()
    with open('C:/Users/A.Mohammadlou/Desktop/software_test/previous_values.txt', 'w') as f:
        f.write(str(i_value))
        f.write('\n')
        f.write(str(distance_y))
        f.write('\n')
        f.write(str(distance_c))
        f.write('\n')
        f.write(str(distance_x))
        f.write('\n')
        f.write(str(distance_x_c))
    update_the_dict(x4,x2)


def create_the_cnc_sub(distance_c, i , substrate_name , code , source_file_path  , free_side , x6):
    directory_path = 'C:/Users/A.Mohammadlou/Desktop/software_test/'
    test_path = 'C:/Users/A.Mohammadlou/Desktop/software_test/test.dxf'
    if os.path.exists(directory_path):
        for filename in os.listdir(directory_path):
            if filename == 'test.dxf':
                file_path = os.path.join(directory_path, filename)
                os.remove(file_path)
    
    target_doc = ezdxf.readfile(tar_path)
    target_models = target_doc.modelspace()
    for entityd in target_models:
        if (entityd.dxf.layer != 'nasb' and entityd.dxf.layer != 'cnc' 
            and entityd.dxf.layer != 'daghi' and entityd.dxf.layer != 'kalaf' 
            and entityd.dxf.layer != 'Defpoints' and entityd.dxf.layer != 'fesharki'
            and entityd.dxf.layer != 'zip' and entityd.dxf.layer != 'tighe zani'):
            entityd.destroy()
        else:
            pass
       
    size1_sw = [ 30.0 , 19.5 ]
    size2_sw = [ 59.0 , 37.5 ]
    size3_sw = [ 59.0 , 39.0 ]
    size1_fn = [ 34.0 , 21.5 ]
    size2_fn = [ 63.0 , 37.5 ]
    size3_fn = [ 75.0 , 44.37]




    def choose_the_best_size(swingo , nexto , firsto):
        if swingo == 'size1':
            if nexto == 'size1':
                if firsto == 'size1':
                    return 'size1'
                elif firsto == 'size2':
                    return 'size2'
                else:
                    return 'size3'
            elif nexto == 'size2':
                if firsto == 'size3':
                    return 'size3'
                else:
                    return 'size2'
            else:
                return 'size3'
        elif swingo == 'size2':
            if nexto == 'size1' or nexto == 'size2':
                if firsto == 'size1' or firsto == 'size2':
                    return 'size2'
                else:
                    return 'size3'
            else:
                return 'size3'
        else:
            return 'size3'

    
    
    


    def sub_pro(layer , cl , free_side):
    
        source_doc = ezdxf.readfile(test_path)
        target_doc = ezdxf.readfile(tar_path)
        source_models = source_doc.modelspace()
        target_models = target_doc.modelspace()

        entity_copy2 = []
    
        for entity in source_models:
    
            
            if entity.dxf.layer == str(layer):
                entity_copy1 = entity.copy()
                entity_copy11 = entity.copy()
                if free_side[1][9] == '':
                    pass
                else:
                    teta = (int(free_side[1][9])/180) * math.pi
                    entity_copy1.rotate_z(teta)
                    #entity_copy1.translate(20000,20000,0)
                    entity_copy11.rotate_z(teta)
                    #entity_copy11.translate(20000,20000,0)

                if str(free_side[1][10]) == 'x':
                    entity_copy1.transform(Matrix44.scale(1,-1,1))
                    entity_copy11.transform(Matrix44.scale(1,-1,1))
                elif str(free_side[1][10]) == 'y':
                    entity_copy1.transform(Matrix44.scale(-1,1,1))
                    entity_copy11.transform(Matrix44.scale(-1,1,1))
                else:
                    pass

                upright(entity_copy1)
                upright(entity_copy11)
                entity_copy2.append(entity_copy11)
                entity_copy1.dxf.color = cl
                ss = target_models.add_entity(entity_copy1)
                
      
           
                target_doc.saveas(tar_path)   
                swing_points = [] 
                #lines = []            
                for entity in target_models:
                
                    if entity.dxf.layer == str(layer):
                        the_swing = entity
                    
                    
                        if the_swing.dxftype() == 'LINE':
                            p_f = []
                            p_s = []
                            p_e = []
                            x , y , z = the_swing.dxf.start
                            p_s.append(x)
                            p_s.append(y)
                            p_s.append(z)
                            x , y , z = the_swing.dxf.end
                            p_e.append(x)
                            p_e.append(y)
                            p_e.append(z)
                            p_f.append(tuple(p_s))
                            p_f.append(tuple(p_e))
                            """
                             check Intersection of lines
                             
                             """
                            
                            
                            swing_points.append(p_f)
                        
                        elif the_swing.dxftype() == 'ARC':
                            p_f = []
                            p_s = []
                            p_e = []
                            x , y , z = the_swing.start_point
                            p_s.append(x)
                            p_s.append(y)
                            p_s.append(z)
                            x , y , z = the_swing.end_point
                            p_e.append(x)
                            p_e.append(y)
                            p_e.append(z)
                            p_f.append(tuple(p_s))
                            p_f.append(tuple(p_e))
                            swing_points.append(p_f)
                            
                        elif the_swing.dxftype() == 'SPLINE':
                            swing_points.append(list(the_swing.control_points))
                    
                        else:
                            swing_points.append(the_swing.get_points())
                    
                    else:
                        pass
                else:
                    pass
        

  
                    
                       
        x_max_min_t = []
        y_max_min_t = []
        x_max_min_f = []
        y_max_min_f = []
        xd_max_min_t = []
        yd_max_min_t = []
    

    
        for j in range(len(swing_points)):
            x_max_min_o = []
            for jj in swing_points[j]:
            
                x_max_min_t.append(jj[0])
                x_max_min_o.append(jj[0])
            
        
            xd_max_min_t.append(x_max_min_o)
        
            x_max_o = max_abs(x_max_min_o)
            x_min_o = min_abs(x_max_min_o)
            x = []
            x.append(x_max_o)
            x.append(x_min_o)
            x_max_min_f.append(x)
       
        for k in range(len(swing_points)):
            y_max_min_o = []
            for kk in swing_points[k]:
                y_max_min_t.append(kk[1])
                y_max_min_o.append(kk[1])
            yd_max_min_t.append(y_max_min_o)
            y_max_o = max_abs(y_max_min_o)
            y_min_o = min_abs(y_max_min_o)
            y = []
            y.append(y_max_o)
            y.append(y_min_o)
            y_max_min_f.append(y)
        x_max_t = max_abs(x_max_min_t)
        y_max_t = max_abs(y_max_min_t)
        x_min_t = min_abs(x_max_min_t)
        y_min_t = min_abs(y_max_min_t)
    
        xy_max_min_f = []
        for ii in range(len(x_max_min_f)):
            xx=[]
            for iii in x_max_min_f[ii]:
                xx.append(iii)
            xy_max_min_f.append(xx)
    
            
    
        for jj in range(len(y_max_min_f)):
            for jjj in y_max_min_f[jj]:
                xy_max_min_f[jj].append(jjj)
            
    

        
        

    
        def for_con():
            global again
            again = False
            for i in range(len(xy_max_min_f)):
                for j in range(len(xy_max_min_f)):
                    if i != j :
                        if ( xy_max_min_f[j][1] >= xy_max_min_f[i][1] and xy_max_min_f[j][0] <= xy_max_min_f[i][0] ) and ( xy_max_min_f[j][3] >= xy_max_min_f[i][3] and xy_max_min_f[j][2] <= xy_max_min_f[i][2] ):
                            x_t = []
                            y_t = []
                            xy_u = []
                            x_t.append(xy_max_min_f[j][0])
                            x_t.append(xy_max_min_f[j][1])
                            x_t.append(xy_max_min_f[i][0])
                            x_t.append(xy_max_min_f[i][1])
                            y_t.append(xy_max_min_f[j][2])
                            y_t.append(xy_max_min_f[j][3])
                            y_t.append(xy_max_min_f[i][2])
                            y_t.append(xy_max_min_f[i][3])
                            x_max_u = max_abs(x_t)
                            x_min_u = min_abs(x_t)
                            y_max_u = max_abs(y_t)
                            y_min_u = min_abs(y_t)
                            xy_u.append(x_max_u)
                            xy_u.append(x_min_u)
                            xy_u.append(y_max_u)
                            xy_u.append(y_min_u)
                            xy_max_min_f.remove(xy_max_min_f[i])
                            if i > j :
                                xy_max_min_f.remove(xy_max_min_f[j])
                            else:
                                xy_max_min_f.remove(xy_max_min_f[j-1])
                        
                            xy_max_min_f.append(xy_u)
                            again = True
                            return
                           
         
        for_con()
        while again == True :
            for_con()
    
    
        
                
            
    
        obj_length = abs (x_max_t - x_min_t) 
        obj_width = abs (y_max_t - y_min_t)
        if layer == 'swing':
            if obj_length <= size1_sw[0] and obj_width <= size1_sw[1] :
                size = 'size1'
            elif obj_length <= size2_sw[0] and obj_width <= size2_sw[1]:
                size = 'size2'
            else : 
                size = 'size3'
        else:
            if obj_length <= size1_fn[0] and obj_width <= size1_fn[1] :
                size = 'size1'
            elif obj_length <= size2_fn[0] and obj_width <= size2_fn[1]:
                size = 'size2'
            else : 
                size = 'size3'
    
        
        obj_pro = [obj_length , obj_width , size , x_min_t , y_min_t , entity_copy2 , xy_max_min_f ,
                   xd_max_min_t , yd_max_min_t]
        if layer == 'yg':
            for obj in target_models:
                if obj.dxf.layer == 'yg':
                    obj.destroy()
                else:
                    pass
            target_doc.saveas(tar_path)
            
        if layer == 'mg':
            for obj in target_models:
                if obj.dxf.layer == 'mg':
                    obj.destroy()
                else:
                    pass
            target_doc.saveas(tar_path)
    
    
        return obj_pro

    def add_the_open_Arcs(x_sw , y_sw , x_nx , y_nx , c_sit , r_p ):
    
        def check_same( x1 , x2 , bl ):
            the_same = []
          
            for i in range(len(x1)):
                bb = False
                for j in range (len(x2)) :
                    for k in range(len(x2[j])):
                        if x1[i] == x2[j][k]:
                            bb = True
                if bb == True:
                    bbl = [bl , i]
                    the_same.append(bbl)
                else:
                    pass

            return the_same
    
    
    

    
        point_added = []
        for bol in range(len(c_sit)):
            if c_sit[bol] == False:
            
                xij_same = check_same(x_sw[bol] , x_nx , bol)
                yij_same = check_same(y_sw[bol] , y_nx , bol)
            
            
                x_add = []
       
                if len(xij_same) > 0 :
                    for a in xij_same:
                        xa = x_sw[a[0]]
                        ya = y_sw[a[0]]
                        x_y_a = []
                        x_y_a.append(xa[a[1]])
                        x_y_a.append(ya[a[1]])
                        x_y_a.append(a[0])
                        x_add.append(x_y_a)
                
            
                y_add = []
                if len(yij_same) > 0 :
                    for b in yij_same:
                        xb = x_sw[b[0]]
                        yb = y_sw[b[0]]
                        x_y_b = []
                        x_y_b.append(xb[b[1]])
                        x_y_b.append(yb[b[1]])
                        x_y_b.append(b[0])
                        y_add.append(x_y_b)
            
            

                if not x_add:
                    pass
                else:
                    for i in x_add:
                        points_a = []
                        x1 = i[0]
                        x2 = i[0]
                        y1 = i[1]+0.25
                        y2 = i[1]-0.25
                        if i[0] < r_p[i[2]][0]:
                            p1 = []
                            p2 = []
                            p1.append(x1)
                            p1.append(y1)
                            p2.append(x2)
                            p2.append(y2)
                            points_a.append(p1)
                            points_a.append(p2)
                        else:
                            p1 = []
                            p2 = []
                            p1.append(x1)
                            p1.append(y1)
                            p2.append(x2)
                            p2.append(y2)
                            points_a.append(p2)
                            points_a.append(p1)
                    
                        point_added.append(points_a)
                
                
                
                if not y_add:
                    pass
                else:
                    for j in y_add:
                        points_a = []
                        x1 = j[0]+0.25
                        x2 = j[0]-0.25
                        y1 = j[1]
                        y2 = j[1]

                        if j[1] > r_p[j[2]][1]:
                            p1 = []
                            p2 = []
                            p1.append(x1)
                            p1.append(y1)
                            p2.append(x2)
                            p2.append(y2)
                            points_a.append(p1)
                            points_a.append(p2)
                        else:
                            p1 = []
                            p2 = []
                            p1.append(x1)
                            p1.append(y1)
                            p2.append(x2)
                            p2.append(y2)
                            points_a.append(p2)
                            points_a.append(p1)
                    
                        point_added.append(points_a)
                    
                
                
            else:
                pass

        return point_added
    
    
    
    
    
    
    
    def create_the_substrates(distance_c, i , substrate_name , code , source_file_path  , free_side , x6):
        get_recently_started_tasks()

        directory_path = 'C:/Users/A.Mohammadlou/Desktop/software_test/'
        if os.path.exists(directory_path):
            for filename in os.listdir(directory_path):
                if filename == 'test.dxf':
                    file_path = os.path.join(directory_path, filename)
                    os.remove(file_path)
        
        
        
        target_doc = ezdxf.readfile(tar_path)
        target_models = target_doc.modelspace()
        for entityd in target_models:
            if (entityd.dxf.layer != 'nasb' and entityd.dxf.layer != 'cnc' 
              and entityd.dxf.layer != 'daghi' and entityd.dxf.layer != 'kalaf' 
               and entityd.dxf.layer != 'Defpoints' and entityd.dxf.layer != 'fesharki'
               and entityd.dxf.layer != 'zip' and entityd.dxf.layer != 'tighe zani'):
                entityd.destroy()
            else:
                pass
               
        swing_next_points = []
        
    
        acad = win32com.client.Dispatch("AutoCAD.Application") 
        acad.Documents.Open(source_file_path)
        
        doc = acad.ActiveDocument
    
        new_layer1 = doc.Layers.Add('first object')
        new_layer2 = doc.Layers.Add('next object')
        new_layer3 = doc.Layers.Add('swing')
        

        
        selSet3 = acad.ActiveDocument.SelectionSets.Add('03')
        selSet3.SelectOnScreen()
        for obj in selSet3:
            obj.layer = 'first object'
        
        if free_side[1][8] == 0:
            selSet2 = acad.ActiveDocument.SelectionSets.Add('02')
            selSet2.SelectOnScreen()
            for obj in selSet2:
                obj.layer = 'next object'
        else:
            pass
    
    
        selSet1 = acad.ActiveDocument.SelectionSets.Add('01')
        selSet1.SelectOnScreen()
        for obj in selSet1:
            obj.layer = 'swing'
            if obj.EntityName == 'AcDbLine':
                close_swing = False
            elif obj.EntityName == 'AcDbArc':
                close_swing = False
            else :
                close_swing = obj.closed
            swing_next_points.append(close_swing)
        
        if free_side[1][5] == 1:
            new_layer4 = doc.Layers.Add('yg')
            selSety = acad.ActiveDocument.SelectionSets.Add('0y')
            selSety.SelectOnScreen()
            for obj in selSety:
                obj.layer = 'yg'
        if free_side[1][6] == 1:
            new_layer5 = doc.Layers.Add('mg')
            selSety = acad.ActiveDocument.SelectionSets.Add('0m')
            selSety.SelectOnScreen()
            for obj in selSety:
                obj.layer = 'mg'
            
        time.process_time()
        doc.Saveas(test_path)
        doc.Close()
        p_swing = sub_pro('swing' , 7 , free_side)
        if free_side[1][8] == 0:
            p_next = sub_pro('next object' , free_side[1][3] , free_side)
        else:
            pass
        p_first = sub_pro('first object' , free_side[0][3] , free_side)
        if free_side[1][5] == 1:
            p_yg = sub_pro('yg' , free_side[1][3] , free_side)
        if free_side[1][6] == 1:
            p_mg = sub_pro('mg' , free_side[1][3] , free_side)
    
    
    
    
    
        #print('xd : ', p_first[7])
        #print('yd : ', p_first[8])
        the_p_r  = []
        for psx , psy in zip(p_swing[7],p_swing[8]) :
            pr_11  = []
            prx_min = min_abs(psx)
            prx_max = max_abs(psx)
            pry_min = min_abs(psy)
            pry_max = max_abs(psy)
            x_center = prx_min + (prx_max - prx_min)/2
            y_center = pry_min + (pry_max - pry_min)/2
            pr_11.append(x_center)
            pr_11.append(y_center)
            the_p_r.append(pr_11)
        
        if free_side[1][8] == 0: 
            point_added = add_the_open_Arcs(p_swing[7] , p_swing[8], p_next[7], p_next[8] , swing_next_points , the_p_r )
            final_size = choose_the_best_size(p_swing[2],p_next[2],p_first[2])      
        else:
            final_size = choose_the_best_size(p_swing[2],p_first[2],p_first[2])
        
    
        target_doc = ezdxf.readfile(tar_path)
        target_models = target_doc.modelspace()
        global c
        c = 1
        if final_size == 'size1':
            length_size1 = 37.75
            width_size1 = 22.5
        
            x_min_swing = p_swing[3]
            y_min_swing = p_swing[4]
            x_max_swing = p_swing[3]+p_swing[0]
            y_max_swing = p_swing[4]+p_swing[1]
        
            reference_point_x11 = p_first[3] - (length_size1 - p_first[0])/2
            reference_point_y11 = p_first[4] - (width_size1 - p_first[1] )/2

            if (reference_point_x11+3.88 < x_min_swing and reference_point_x11+33.88 > x_max_swing 
                and reference_point_y11+3 < y_min_swing and reference_point_y11+22.5 > y_max_swing):
                reference_point_x1 = reference_point_x11
                reference_point_y1 = reference_point_y11
            
            elif reference_point_y11+3 < y_min_swing and reference_point_y11+22.5 > y_max_swing :
                reference_point_y1 = reference_point_y11
            
            elif reference_point_x11+3.88 < x_min_swing and reference_point_x11+33.88 > x_max_swing:
                reference_point_x1 = reference_point_x11 
            
            if reference_point_x11+3.88 > x_min_swing:
                diff_sx = reference_point_x11 + 3.88 - x_min_swing
                reference_point_x1 = reference_point_x11 - diff_sx - 1
            
            elif reference_point_x11 + 33.88 < x_max_swing:
                diff_sx = x_max_swing - (reference_point_x11+33.88)
                reference_point_x1 = reference_point_x11 + diff_sx + 1

            if reference_point_y11 + 3 > y_min_swing:
                diff_sy = reference_point_y11+3 - y_min_swing
                reference_point_y1 = reference_point_y11 - diff_sy - 1

            elif reference_point_y11 + 22.5 < y_max_swing:
                diff_sy = y_max_swing - (reference_point_y11 + 22.5)
                reference_point_y1 = reference_point_y11 + diff_sy + 1

            ####
            if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                pass
                        
            elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==0:
                dff = (reference_point_x1 + length_size1) - (p_first[3] + p_first[0]) 
                reference_point_x1 = reference_point_x1 - dff
                reference_point_y1 = reference_point_y1 
                    
                        
            elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==0:
                dff = (reference_point_x1) - (p_first[3])
                reference_point_x1 = reference_point_x1 - dff
                reference_point_y1 = reference_point_y1
                   
                       
            elif free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==1:
                dff = p_first[4] - reference_point_y1
                reference_point_x1 = reference_point_x1
                reference_point_y1 = reference_point_y1 + dff
                  
                        
            elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==0:
                reference_point_x1 = reference_point_x1
                reference_point_y1 = reference_point_y1

                    
                       
            elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==1:
                dffy = p_first[4] - reference_point_y1
                dffx = (reference_point_x1 + length_size1) - (p_first[3] + p_first[0])
                reference_point_x1 = reference_point_x1 - dffx
                reference_point_y1 = reference_point_y1 + dffy
                   
                    
            elif free_side[0][0]== 0 and free_side[0][1]==1 and free_side[0][2]==1:
                dffy = p_first[4] - reference_point_y1
                dffx = (reference_point_x1) - (p_first[3])
                reference_point_x1 = reference_point_x1 - dffx
                reference_point_y1 = reference_point_y1 + dffy
                   
                       
            else:
                dff = p_first[4] - reference_point_y1
                reference_point_x1 = reference_point_x1
                reference_point_y1 = reference_point_y1 + dff
                
            ###
        
        
            origin = Vec3(reference_point_x1,reference_point_y1)
            rect_1 = target_models.add_lwpolyline(forms.translate(forms.box(length_size1 , width_size1+3),
                                    origin), close = True , dxfattribs= {'color': free_side[0][3]})
            rect_0 = rect_1.copy()
            rect_0.translate((length_size1)+ 5 , 0 ,0)
            rect_0.dxf.color = 5
            target_models.add_entity(rect_0)
            
            ####!!!@@@#####
            
            if free_side[1][8] == 1:
                if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                    xy_pc = []
                    for idc in range(len(p_first[7])):
                        points = list(zip(p_first[7][idc], p_first[8][idc]))
                        min_x1 = min(points, key=lambda p: p[0])  
                        min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                        p1d = min(min_x_points1, key=lambda p: p[1])
                        
                        max_y11 = min(points, key=lambda p: p[1])  
                        max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                        p2d = min(max_y_points11, key=lambda p: p[0])
                        
                        if round(p1d[0], 5) != round(p2d[0] , 5) and round(p1d[1] , 5) != round(p2d[1] , 5):
                            lg = abs(p2d[0] - p1d[0])
                            wd = abs(p2d[1] - p1d[1])
                            if wd >= 1 :
                                ps1 = p1d
                                pe1 = (p1d[0]-1, p1d[1])
                                l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps2 = pe1
                                pe2 = (pe1[0],pe1[1]-1)
                                l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps3 = pe2
                                pe3 = p2d
                                l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                            
                            elif wd < 1 :
                                df = (1 - wd)/2
                                
                                ps1 = p1d
                                pe1 = (p1d[0] , p1d[1]+df)
                                l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps2 = pe1
                                pe2 = (pe1[0] - 1, pe1[1])
                                l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps3 = pe2
                                pe3 = (pe2[0] , pe2[1] - 1)
                                l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps4 = pe3
                                pe4 = (p2d[0] , ps4[1])
                                l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps5 = pe4
                                pe5 = p2d
                                l5 = target_models.add_line( ps5 ,pe5 , dxfattribs = {'color': free_side[0][3]} )
                        
                        else:  
                            ps1 = (p1d[0] , p1d[1]+0.5)
                            pe1 = (ps1[0]-0.5 , ps1[1])
                            l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                            
                            ps2 = pe1
                            pe2 = (pe1[0],pe1[1]-1)
                            l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                            
                            ps3 = pe2
                            pe3 = (ps3[0]+1 , ps3[1])
                            l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                            
                            ps4 = pe3
                            pe4 = (ps4[0] , ps4[1]+0.5)
                            l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[0][3]} )
            
                                
                            
                        
                        
                    
                    
                
            
            if free_side[1][8] == 0:
            
                ###### add sight ######
            
                if free_side[1][7] == 1:
                    xy_pc = []
                    for iac in range(len(p_first[7])):
                        points = list(zip(p_first[7][iac], p_first[8][iac]))
                        min_x1 = min(points, key=lambda p: p[0])  
                        min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                        p1 = max(min_x_points1, key=lambda p: p[1]) 
                        max_x2 = max(points, key=lambda p: p[0])
                        max_x_points2 = [p for p in points if p[0] == max_x2[0]]
                        p2 = max(max_x_points2, key=lambda p: p[1])    
                        max_y11 = max(points, key=lambda p: p[1])  
                        max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                        p11 = min(max_y_points11, key=lambda p: p[0])
                        max_y22 = max(points, key=lambda p: p[1])
                        max_y_points22 = [p for p in points if p[1] == max_y22[1]]
                        p22 = max(max_y_points22, key=lambda p: p[0])
                        if round(p1[0] , 5) != round(p11[0], 5) and round(p1[1] , 5) != round(p11[1], 5):
                            lg1 = p11[0] - p1[0]
                            wd1 = p11[1] - p1[1]
                            cp1 = (p11[0]-lg1/2 , p11[1]-wd1/2)
                        else:
                            cp1 = p11
        
                        if round(p2[0] , 5) != round(p22[0] , 5) and round(p2[1] , 5) != round(p22[1] , 5):
                            lg2 = p2[0] - p22[0]
                            wd2 = p22[1] - p2[1]
                            cp2 = (p22[0]+lg2/2 , p22[1]-wd2/2)
                        else:
                            cp2 = p22

                        l = []
                        l.append(cp1)
                        l.append(cp2)
                        xy_pc.append(l)
            
                    for cs in xy_pc :
                        circle_sight_L = target_models.add_circle(cs[0] , radius=2 , dxfattribs={'color':free_side[1][3]})
                        circle_sight_R = target_models.add_circle(cs[1] , radius=2 , dxfattribs={'color':free_side[1][3]})
                        circle_sight_L.translate(-(length_size1+ 5) , 0 ,0)
                        circle_sight_R.translate(-(length_size1+ 5) , 0 ,0)
                else:
                    pass
            else:
                pass
   
                
            for swing_rec in p_swing[6]:
                lengths0 = abs(swing_rec[0] - swing_rec[1])+0.4
                widths0 = abs(swing_rec[2] - swing_rec[3])+0.4
                reference_point_x_s = swing_rec[1] - 0.2
                reference_point_y_s = swing_rec[3] - 0.2
            
                origin = Vec3(reference_point_x_s , reference_point_y_s)
                rect_s = target_models.add_lwpolyline(forms.translate(forms.box(lengths0, widths0), origin), 
                                                 close = True , dxfattribs={'color':5})
                
                rect_s.translate((length_size1)+5 , 0 , 0)

                
            
            
        
            if free_side[1][8] == 0 :
                #if p_next[0]+8 >= length_size1 :
                #    length_rec_next = length_size1
                #else:
                length_rec_next = p_next[0] #+ 8
                width_rec_next = p_next[1]
        
        

        
            
                reference_point_x_3 = p_next[3] #- (length_rec_next - p_next[0])/2
                reference_point_y_3 = p_next[4] #- 4
        
                the_diff = (reference_point_y1 + width_size1) - (reference_point_y_3 + 4 + width_rec_next)
        
        
                width_rec_next = width_rec_next + the_diff

                #if p_next[1] + 4 >= width_size1 :
                #    width_rec_next = width_size1
                #else:
                #    width_rec_next = width_rec_next + 4
                
                
                H_f1 = 4
                if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
                    length_rec_next = length_rec_next + 2*H_f1
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3 - H_f1
                  
                        
                elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3 - H_f1
                    
                        
                elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3 - H_f1
                   
                       
                elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
                    length_rec_next = length_rec_next + 2* H_f1
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3
                  
                        
                elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
                    length_rec_next = length_rec_next
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3 - H_f1
                    
                       
                elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3
                   
                    
                elif free_side[1][0]== 0 and free_side[1][1]==1 and free_side[1][2]==1:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3
                   
                       
                else:
                    length_rec_next = length_rec_next
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3
            

        
        
                origin_3 = Vec3(reference_point_x_3 , reference_point_y_3)
                rect_3 = target_models.add_lwpolyline(forms.translate(forms.box(length_rec_next,width_rec_next+3),
                                        origin_3),close=True , dxfattribs = {'color': free_side[1][3]})
                ##### add yaragh ####
            
                if free_side[1][5] == 1:
                    for yi in range(len(p_yg[7])):
                        xmx = max_abs(p_yg[7][yi])
                        xmn = min_abs(p_yg[7][yi])
                        ymx = max_abs(p_yg[8][yi])
                        ymn = min_abs(p_yg[8][yi])
                        lg = xmx - xmn + 4
                        wd = ymx - ymn + 2.3
                        rfx = xmn - 2
                        rfy = ymn - 0.3
                        originy = Vec3(rfx,rfy)
                        rcy = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd) , originy),
                                    close=True , dxfattribs = {'color': free_side[1][3], 'layer':'next object'})
                else:
                    pass
            
                #### add maghzi ####
            
                if free_side[1][6] == 1:
                    for mi in range(len(p_mg[7])):
                        xmx = max_abs(p_mg[7][mi])
                        xmn = min_abs(p_mg[7][mi])
                        ymx = max_abs(p_mg[8][mi])
                        ymn = min_abs(p_mg[8][mi])
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn - 0.5
                        rfy = ymn - 0.5
                        originy = Vec3(rfx,rfy)
                        rcm = target_models.add_lwpolyline(forms.translate(forms.box(lg,wd) , originy),
                                close = True , dxfattribs= {'color': 3})
                        txt0=target_models.add_text(str(i+1), height = 0.3 , 
                           dxfattribs = {'style':'TXT','color':4}).set_placement((
                               rfx+0.625,rfy+0.625),align = TextEntityAlignment.LEFT)
                    
                        rcm.translate(-(2*length_size1+5) , 0 , 0)
                        txt0.translate(-(2*length_size1+5) , 0 , 0)
                        mg = p_mg[5][mi].copy()
                        mg.dxf.color = 7
                        mg.translate(-(2*length_size1+5) , 0 , 0)
                        target_models.add_entity(mg)
                        mg1 = mg.copy()
                        txt1 = txt0.copy()
                        mg1.translate(0 , -wd , 0)
                        mg1.dxf.color = 1
                        txt1.translate(0 , -wd , 0)
                        target_models.add_entity(mg1)
                        target_models.add_entity(txt1)
                        mg2=mg1.copy()
                        txt2=txt1.copy()
                        mg2.translate(0 , -wd+1 , 0)
                        txt2.translate(0 , -wd+1 , 0)
                        target_models.add_entity(mg2)
                        target_models.add_entity(txt2)
                        originy = Vec3(rfx + lg/2 - 2 , rfy + wd/2 - 0.25)
                        rcmm = target_models.add_lwpolyline(forms.translate(forms.box(4,0.5) , originy),
                                close = True , dxfattribs= {'color': 7})
                        rcmm.translate(-(2*length_size1+5) , -wd , 0)
                        txt00 = target_models.add_text(str(i+1), height = 0.3 , 
                           dxfattribs = {'style':'TXT','color':4}).set_placement((
                               rfx + lg/2 - 1.5 , rfy + wd/2 - 0.125),align = TextEntityAlignment.LEFT)
                        txt00.translate(-(2*length_size1+5) , -wd , 0)
                        r1 = rcmm.copy()
                        r1.dxf.color = 1
                        r1.translate(0,-2*wd + 2 ,0)
                        txt00.translate(0,-2*wd + 2,0)
                        target_models.add_entity(r1)
                        r2 = r1.copy()
                        txt11=txt00.copy()
                        r2.translate(4 , 0 , 0)
                        txt11.translate(4 , 0 , 0)
                        target_models.add_entity(r2)
                        target_models.add_entity(txt11)
                        r3 = r2.copy()
                        txt22=txt11.copy()
                        r3.translate(4 , 0 , 0)
                        txt22.translate(4 , 0 , 0)
                        target_models.add_entity(r3)
                        target_models.add_entity(txt22)
                else:
                    pass
            
        
                PL_s_x1 = reference_point_x_3
                PL_s_y1 = reference_point_y_3 + width_rec_next
                PL_e_x1 = reference_point_x_3 + length_rec_next
        
        
        
                line_1_horizontal = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_e_x1 , PL_s_y1) , 
                                       dxfattribs = {'color': 7} )
                line_3_horizontal = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_e_x1 , PL_s_y1) , 
                                       dxfattribs = {'color': free_side[1][3]} )
                line_1_vertical = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_s_x1 , PL_s_y1+3) , 
                                       dxfattribs = {'color': 7} )
                line_2_vertical = target_models.add_line( (PL_s_x1 + length_rec_next , PL_s_y1) ,
                                                 (PL_s_x1 + length_rec_next , PL_s_y1+3) , 
                                                 dxfattribs = {'color': 7})
        
        
                rect_3.translate(-(length_size1+5) , 0 , 0)
                line_3_horizontal.translate(-(length_size1 +5 ), 0 , 0)
        
                if not point_added:
                    pass
                else:
                    for ar in point_added:
                        p1 = ezdxf.math.Vec2(ar[0])
                        p2 = ezdxf.math.Vec2(ar[1])

                        arc = ConstructionArc.from_2p_radius(p1 , p2 , 0.2501 , True, True)
                        arc.add_to_layout(target_models , dxfattribs={'layer':'next object' ,
                                                                  'color':free_side[1][3]})
        
                
        
                for co_entity in p_first[5]:
                    c_entity = co_entity.copy()
                    c_entity.translate((length_size1)+5 , 0 , 0)
                    c_entity.dxf.color = 7
                    target_models.add_entity(c_entity)
            
            
                target_models.add_text(str(substrate_name), height = 0.6 , 
                   dxfattribs = {'layer':'next object', 'style':'TXT','color':4}).set_placement((
                   length_rec_next/2 + reference_point_x_3,
                   reference_point_y_3+2) ,align = TextEntityAlignment.CENTER)
                target_models.add_text('code: '+str(code) , height = 0.6 ,
                   dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                   length_rec_next/2 + reference_point_x_3 , reference_point_y_3+1) , align = TextEntityAlignment.CENTER)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                        reference_point_x_3+1,reference_point_y_3+1) ,align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs ={'layer':'next object','style':'TXT','color':4}).set_placement((
                        reference_point_x_3+1,reference_point_y_3+width_rec_next+1) , align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs = {'layer':'first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + 1 ,reference_point_y1+1) ,align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs= {'layer':' first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + length_size1 + 6 ,reference_point_y1+1) ,
                        align = TextEntityAlignment.LEFT)            
            
                for txt in p_next[6]:
                    x = (txt[0] - txt[1])/2 + txt[1]
                    y = txt[3] + 1
                    target_models.add_text(str(i+1) , height=1 , 
                                  dxfattribs={'layer':'next object' , 'style':'TXT','color':4}).set_placement((
                                  x , y), align = TextEntityAlignment.LEFT)
            
            
            else:
                target_models.add_text(str(substrate_name), height = 0.6 , 
                   dxfattribs = {'layer':'next object', 'style':'TXT','color':4}).set_placement((
                   3*length_size1/2 + reference_point_x1+5,
                   reference_point_y1+2) ,align = TextEntityAlignment.CENTER)
                target_models.add_text('code: '+str(code) , height = 0.6 ,
                   dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                   3*length_size1/2 + reference_point_x1+5 , reference_point_y1+1) , align = TextEntityAlignment.CENTER)
                #target_models.add_text(str(i+1), height = 1, 
                #        dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                #        reference_point_x_3+1,reference_point_y_3+1) ,align = TextEntityAlignment.LEFT)
                #target_models.add_text(str(i+1), height = 1, 
                #        dxfattribs ={'layer':'next object','style':'TXT','color':4}).set_placement((
                #        reference_point_x_3+1,reference_point_y_3+width_rec_next+1) , align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs = {'layer':'first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + 1 ,reference_point_y1+1) ,align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs= {'layer':' first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + length_size1 + 6 ,reference_point_y1+1) ,
                        align = TextEntityAlignment.LEFT) 
                for txt in p_first[6]:
                    x = (txt[0] - txt[1])/2 + txt[1] + length_size1+5
                    y = txt[3] + 1
                    target_models.add_text(str(i+1) , height=1 , 
                                  dxfattribs={'layer':'next object' , 'style':'TXT','color':4}).set_placement((
                                  x , y), align = TextEntityAlignment.LEFT)
                
            
            ####$$$$$$#####
            
            for entity in target_models:
                if entity.dxf.layer == 'next object':
                    entity.translate(-(length_size1+5) ,  0 , 0 )
                    
                elif entity.dxf.layer == 'swing':
                    if free_side[1][8] == 0:
                        entity.translate(-(length_size1+5) ,  0 , 0 )
                    else:
                        pass
                        
            
            
            
            
            
            
            for en in target_models:
                if en.dxf.layer != 'cnc' and en.dxf.layer != 'nasb': 
                    en.translate(-(reference_point_x1) + 2*length_size1 + 5 + x6,
                                 -(reference_point_y1)+distance_c, 0)
                else:
                    pass
                if en.dxf.layer == 'nasb':
                    pass
                else:
                    en.dxf.layer = 'cnc'

             
           
            if free_side[1][8] == 0:
                if free_side[1][4] == 1 :
                    if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                        
                            dd = dd - lg 
                        
                    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                 close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                               dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                               -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 1
                            rfx = xmn - 1
                            rfy = ymn
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 , distance_c +0.5 + wd) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg

                    elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                  
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 2
                        rfx = xmn
                        rfy = ymn - 1
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                     
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd - 0.5) ,align = TextEntityAlignment.LEFT)
                   
                
                    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn - 1
                        rfy = ymn 
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    
                    elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    else:
                        pass
                        
                        
            else:
                if free_side[0][4] == 1 :
                    if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                            
                            points = list(zip(p_first[7][n] , p_first[8][n] ))
                            min_x1 = min(points, key=lambda p: p[0])  
                            min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                            p1d = min(min_x_points1, key=lambda p: p[1])
                            p1d = (p1d[0]-xmn - 10 + dd+x6 , p1d[1]-ymn + distance_c+1)
                        
                            max_y11 = min(points, key=lambda p: p[1])  
                            max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                            p2d = min(max_y_points11, key=lambda p: p[0])
                            p2d = (p2d[0]-xmn - 10 + dd+x6 , p2d[1]-ymn + distance_c+1)
                                
                        
                            if round(p1d[0], 5) != round(p2d[0] , 5) and round(p1d[1] , 5) != round(p2d[1] , 5):
                                lg1 = abs(p2d[0] - p1d[0])
                                wd1 = abs(p2d[1] - p1d[1])
                                if wd1 >= 1 :
                                    ps1 = p1d
                                    pe1 = (p1d[0]-1, p1d[1])
                                    l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps2 = pe1
                                    pe2 = (pe1[0],pe1[1]-1)
                                    l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps3 = pe2
                                    pe3 = p2d
                                    l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                            
                                elif wd1 < 1 :
                                    df = (1 - wd1)/2
                                
                                    ps1 = p1d
                                    pe1 = (p1d[0] , p1d[1]+df)
                                    l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps2 = pe1
                                    pe2 = (pe1[0] - 1, pe1[1])
                                    l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps3 = pe2
                                    pe3 = (pe2[0] , pe2[1] - 1)
                                    l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps4 = pe3
                                    pe4 = (p2d[0] , ps4[1])
                                    l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps5 = pe4
                                    pe5 = p2d
                                    l5 = target_models.add_line( ps5 ,pe5 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                        
                            else:  
                                ps1 = (p1d[0] , p1d[1]+0.5)
                                pe1 = (ps1[0]-0.5 , ps1[1])
                                pl1 = [ps1,pe1]
                                l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                ps2 = pe1
                                pe2 = (pe1[0],pe1[1]-1)
                                l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                ps3 = pe2
                                pe3 = (ps3[0]+1 , ps3[1])
                                l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                ps4 = pe3
                                pe4 = (ps4[0] , ps4[1]+0.5)
                                l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                            
                            
                            
                            
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -11 + lg/2 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                        
                            dd = dd - lg 
                        
                    elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==0:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                 close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                               dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                               -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==0:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==1:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 1
                            rfx = xmn - 1
                            rfy = ymn
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 , distance_c +0.5 + wd) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg

                    elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==0:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                  
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 2
                        rfx = xmn
                        rfy = ymn - 1
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                     
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd - 0.5) ,align = TextEntityAlignment.LEFT)
                   
                
                    elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn - 1
                        rfy = ymn 
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    
                    elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    else:
                        pass
                

            
            target_doc.saveas(tar_path)
            distance_c = distance_c + 25.5
                        
                        
                        
                        

                
                    
    

                    
                    
                    
        elif final_size == 'size2':
            length_size2 = 71
            width_size2 = 42.37
        
            x_min_swing = p_swing[3]
            y_min_swing = p_swing[4]
            x_max_swing = p_swing[3]+p_swing[0]
            y_max_swing = p_swing[4]+p_swing[1]

            reference_point_x11 = p_first[3] - (length_size2 - p_first[0])/2
            reference_point_y11 = p_first[4] - (width_size2 - p_first[1] )/2

            if reference_point_x11+6 < x_min_swing and reference_point_x11+65 > x_max_swing and reference_point_y11+5.94 < y_min_swing and reference_point_y11+43.44 > y_max_swing:
                reference_point_x1 = reference_point_x11
                reference_point_y1 = reference_point_y11
            
            elif reference_point_y11+5.94 < y_min_swing and reference_point_y11+43.44 > y_max_swing :
                reference_point_y1 = reference_point_y11
            
            elif reference_point_x11+6 < x_min_swing and reference_point_x11+65 > x_max_swing:
                reference_point_x1 = reference_point_x11

            if reference_point_x11+6 > x_min_swing:
                diff_sx = reference_point_x11 + 6 - x_min_swing
                reference_point_x1 = reference_point_x11 - diff_sx - 1
            
            elif reference_point_x11 + 65 < x_max_swing:
                diff_sx = x_max_swing - (reference_point_x11+65)
                reference_point_x1 = reference_point_x11 + diff_sx + 1

            if reference_point_y11 + 5.94 > y_min_swing:
                diff_sy = reference_point_y11+5.94 - y_min_swing
                reference_point_y1 = reference_point_y11 - diff_sy - 1

            elif reference_point_y11 + 43.44 < y_max_swing:
                diff_sy = y_max_swing - (reference_point_y11 + 43.44)
                reference_point_y1 = reference_point_y11 + diff_sy + 1
            ###
            
            if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                pass
                        
            elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==0:
                dff = (reference_point_x1 + length_size2) - (p_first[3] + p_first[0]) 
                reference_point_x1 = reference_point_x1 - dff
                reference_point_y1 = reference_point_y1 
                    
                        
            elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==0:
                dff = (reference_point_x1) - (p_first[3])
                reference_point_x1 = reference_point_x1 - dff
                reference_point_y1 = reference_point_y1
                   
                       
            elif free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==1:
                dff = p_first[4] - reference_point_y1
                reference_point_x1 = reference_point_x1
                reference_point_y1 = reference_point_y1 + dff
                  
                        
            elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==0:
                reference_point_x1 = reference_point_x1
                reference_point_y1 = reference_point_y1

                    
                       
            elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==1:
                dffy = p_first[4] - reference_point_y1
                dffx = (reference_point_x1 + length_size2) - (p_first[3] + p_first[0])
                reference_point_x1 = reference_point_x1 - dffx
                reference_point_y1 = reference_point_y1 + dffy
                   
                    
            elif free_side[0][0]== 0 and free_side[0][1]==1 and free_side[0][2]==1:
                dffy = p_first[4] - reference_point_y1
                dffx = (reference_point_x1) - (p_first[3])
                reference_point_x1 = reference_point_x1 - dffx
                reference_point_y1 = reference_point_y1 + dffy
                   
                       
            else:
                dff = p_first[4] - reference_point_y1
                reference_point_x1 = reference_point_x1
                reference_point_y1 = reference_point_y1 + dff
            ###

        
            c1 = (reference_point_x1+1.75 , reference_point_y1+11.5)
            c2 = (reference_point_x1+1.75 , reference_point_y1+23.5)
            c3 = (reference_point_x1+1.6976 , reference_point_y1+36.0512)
            c4 = (reference_point_x1+69.2476 , reference_point_y1+36.0512)
            c5 = (reference_point_x1+69.25 , reference_point_y1+23.5)
            c6 = (reference_point_x1+69.25 , reference_point_y1+11.5)
            c7 = (reference_point_x1+41.5 , reference_point_y1+1.5862)
            c8 = (reference_point_x1+29.5 , reference_point_y1+1.5862)
        
        
            origin = Vec3(reference_point_x1,reference_point_y1)
            rect_1 = target_models.add_lwpolyline(forms.translate(forms.box(length_size2 , width_size2+3),
                                    origin), close = True , dxfattribs= {'color': free_side[0][3]})
            circle_11 = target_models.add_circle(c1 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
            circle_21 = target_models.add_circle(c2 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
            circle_31 = target_models.add_circle(c3 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
            circle_41 = target_models.add_circle(c4 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
            circle_51 = target_models.add_circle(c5 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
            circle_61 = target_models.add_circle(c6 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
            circle_71 = target_models.add_circle(c7 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
            circle_81 = target_models.add_circle(c8 , radius=0.6095 , dxfattribs={'color':free_side[0][3]})
        
            rect_0 = rect_1.copy()
            circle_10 = circle_11.copy()
            circle_20 = circle_21.copy()
            circle_30 = circle_31.copy()
            circle_40 = circle_41.copy()
            circle_50 = circle_51.copy()
            circle_60 = circle_61.copy()
            circle_70 = circle_71.copy()
            circle_80 = circle_81.copy()
        
        
            rect_0.translate(length_size2+5, 0, 0)
            rect_0.dxf.color = 5
            circle_10.translate(length_size2+5, 0, 0)
            circle_10.dxf.color = 5
            circle_20.translate(length_size2+5, 0, 0)
            circle_20.dxf.color = 5
            circle_30.translate(length_size2+5, 0, 0)
            circle_30.dxf.color = 5
            circle_40.translate(length_size2+5, 0, 0)
            circle_40.dxf.color = 5
            circle_50.translate(length_size2+5, 0, 0)
            circle_50.dxf.color = 5
            circle_60.translate(length_size2+5, 0, 0)
            circle_60.dxf.color = 5
            circle_70.translate(length_size2+5, 0, 0)
            circle_70.dxf.color = 5
            circle_80.translate(length_size2+5, 0, 0)
            circle_80.dxf.color = 5
        
    
            target_models.add_entity(rect_0)
            target_models.add_entity(circle_10)
            target_models.add_entity(circle_20)
            target_models.add_entity(circle_30)
            target_models.add_entity(circle_40)
            target_models.add_entity(circle_50)
            target_models.add_entity(circle_60)
            target_models.add_entity(circle_70)
            target_models.add_entity(circle_80)
            
            
            
            
            if free_side[1][8] == 1:
                if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                    xy_pc = []
                    for idc in range(len(p_first[7])):
                        points = list(zip(p_first[7][idc], p_first[8][idc]))
                        min_x1 = min(points, key=lambda p: p[0])  
                        min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                        p1d = min(min_x_points1, key=lambda p: p[1])
                        
                        max_y11 = min(points, key=lambda p: p[1])  
                        max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                        p2d = min(max_y_points11, key=lambda p: p[0])
                        
                        if round(p1d[0], 5) != round(p2d[0] , 5) and round(p1d[1] , 5) != round(p2d[1] , 5):
                            lg = abs(p2d[0] - p1d[0])
                            wd = abs(p2d[1] - p1d[1])
                            if wd >= 1 :
                                ps1 = p1d
                                pe1 = (p1d[0]-1, p1d[1])
                                l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps2 = pe1
                                pe2 = (pe1[0],pe1[1]-1)
                                l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps3 = pe2
                                pe3 = p2d
                                l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                            
                            elif wd < 1 :
                                df = (1 - wd)/2
                                
                                ps1 = p1d
                                pe1 = (p1d[0] , p1d[1]+df)
                                l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps2 = pe1
                                pe2 = (pe1[0] - 1, pe1[1])
                                l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps3 = pe2
                                pe3 = (pe2[0] , pe2[1] - 1)
                                l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps4 = pe3
                                pe4 = (p2d[0] , ps4[1])
                                l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[0][3]} )
                                
                                ps5 = pe4
                                pe5 = p2d
                                l5 = target_models.add_line( ps5 ,pe5 , dxfattribs = {'color': free_side[0][3]} )
                        
                        else:  
                            ps1 = (p1d[0] , p1d[1]+0.5)
                            pe1 = (ps1[0]-0.5 , ps1[1])
                            l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                            
                            ps2 = pe1
                            pe2 = (pe1[0],pe1[1]-1)
                            l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                            
                            ps3 = pe2
                            pe3 = (ps3[0]+1 , ps3[1])
                            l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                            
                            ps4 = pe3
                            pe4 = (ps4[0] , ps4[1]+0.5)
                            l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[0][3]} )
            
            
            
            ###### add sight ######
            if free_side[1][8] ==  0:
                if free_side[1][7] == 1:
                    xy_pc = []
                    for iac in range(len(p_first[7])):
                        points = list(zip(p_first[7][iac], p_first[8][iac]))
                        min_x1 = min(points, key=lambda p: p[0])  
                        min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                        p1 = max(min_x_points1, key=lambda p: p[1]) 
    
                        max_x2 = max(points, key=lambda p: p[0])
                        max_x_points2 = [p for p in points if p[0] == max_x2[0]]
                        p2 = max(max_x_points2, key=lambda p: p[1])
    
                        max_y11 = max(points, key=lambda p: p[1])  
                        max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                        p11 = min(max_y_points11, key=lambda p: p[0])
    
                        max_y22 = max(points, key=lambda p: p[1])
                        max_y_points22 = [p for p in points if p[1] == max_y22[1]]
                        p22 = max(max_y_points22, key=lambda p: p[0])
    
                        if round(p1[0] , 5) != round(p11[0], 5) and round(p1[1] , 5) != round(p11[1], 5):
                            lg1 = p11[0] - p1[0]
                            wd1 = p11[1] - p1[1]
                            cp1 = (p11[0]-lg1/2 , p11[1]-wd1/2)
                        else:
                            cp1 = p11
        
                        if round(p2[0] , 5) != round(p22[0] , 5) and round(p2[1] , 5) != round(p22[1] , 5):
                            lg2 = p2[0] - p22[0]
                            wd2 = p22[1] - p2[1]
                            cp2 = (p22[0]+lg2/2 , p22[1]-wd2/2)
                        else:
                            cp2 = p22

                        l = []
                        l.append(cp1)
                        l.append(cp2)
                        xy_pc.append(l)
            
                    for cs in xy_pc :
                        circle_sight_L = target_models.add_circle(cs[0] , radius=2 , dxfattribs={'color':free_side[1][3]})
                        circle_sight_R = target_models.add_circle(cs[1] , radius=2 , dxfattribs={'color':free_side[1][3]})
                        circle_sight_L.translate(-(length_size2+ 5) , 0 ,0)
                        circle_sight_R.translate(-(length_size2+ 5) , 0 ,0)
                else:
                    pass
            else:
                pass
            
        
            for swing_rec in p_swing[6]:
                lengths0 = abs(swing_rec[0] - swing_rec[1])+0.4
                widths0 = abs(swing_rec[2] - swing_rec[3])+0.4
                reference_point_x_s = swing_rec[1] - 0.2
                reference_point_y_s = swing_rec[3] - 0.2
            
                origin = Vec3(reference_point_x_s , reference_point_y_s)
                rect_s = target_models.add_lwpolyline(forms.translate(forms.box(lengths0, widths0), origin), 
                                                 close = True , dxfattribs={'color':5})
                rect_s.translate((length_size2)+5 , 0 , 0)
            
            if free_side[1][8] == 0 :   
                length_rec_next = p_next[0] #+ 8
                width_rec_next = p_next[1]
                reference_point_x_3 = p_next[3] - (length_rec_next - p_next[0])/2
                reference_point_y_3 = p_next[4] #- 4
                the_diff = (reference_point_y1 + width_size2) - (reference_point_y_3 + 4 + width_rec_next)
                width_rec_next = width_rec_next + the_diff
                H_f1 = 4
                if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
                    length_rec_next = length_rec_next + 2*H_f1
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3 - H_f1
                  
                        
                elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3 - H_f1
                    
                        
                elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3 - H_f1
                   
                       
                elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
                    length_rec_next = length_rec_next + 2* H_f1
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3
                  
                        
                elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
                    length_rec_next = length_rec_next
                    width_rec_next = width_rec_next + 2*H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3 - H_f1
                    
                       
                elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3 - H_f1
                    reference_point_y_3 = reference_point_y_3
                   
                    
                elif free_side[1][0]== 0 and free_side[1][1]==1 and free_side[1][2]==1:
                    length_rec_next = length_rec_next + H_f1
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3
                   
                       
                else:
                    length_rec_next = length_rec_next
                    width_rec_next = width_rec_next + H_f1
                    reference_point_x_3 = reference_point_x_3
                    reference_point_y_3 = reference_point_y_3 
        
        
        
        
                origin_3 = Vec3(reference_point_x_3 , reference_point_y_3)
                rect_3 = target_models.add_lwpolyline(forms.translate(forms.box(length_rec_next,width_rec_next+3),origin_3),
                                                   close=True , dxfattribs = {'color': free_side[1][3]})
        
            
                #### add yaragh #### 
                if free_side[1][5] == 1:
                    for yi in range(len(p_yg[7])):
                        xmx = max_abs(p_yg[7][yi])
                        xmn = min_abs(p_yg[7][yi])
                        ymx = max_abs(p_yg[8][yi])
                        ymn = min_abs(p_yg[8][yi])
                        lg =  xmx - xmn + 4 
                        wd = ymx - ymn + 2.3 
                        rfx = xmn - 2
                        rfy = ymn - 0.3
                        originy = Vec3(rfx,rfy)
                        rcy = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd) , originy),
                                    close=True , dxfattribs = {'color': free_side[1][3], 'layer':'next object'})
                else:
                    pass
            
            
                #### add maghzi ####
            
                if free_side[1][6] == 1:
                    for mi in range(len(p_mg[7])):
                        xmx = max_abs(p_mg[7][mi])
                        xmn = min_abs(p_mg[7][mi])
                        ymx = max_abs(p_mg[8][mi])
                        ymn = min_abs(p_mg[8][mi])
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn - 0.5
                        rfy = ymn - 0.5
                        originy = Vec3(rfx,rfy)
                        rcm = target_models.add_lwpolyline(forms.translate(forms.box(lg,wd) , originy),
                                close = True , dxfattribs= {'color': 3})
                        txt0=target_models.add_text(str(i+1), height = 0.3 , 
                             dxfattribs = {'style':'TXT','color':4}).set_placement((
                               rfx+0.625,rfy+0.625),align = TextEntityAlignment.LEFT)
                    
                        rcm.translate(-(2*length_size2+5) , 0 , 0)
                        txt0.translate(-(2*length_size2+5) , 0 , 0)
                        mg = p_mg[5][mi].copy()
                        mg.dxf.color = 7
                        mg.translate(-(2*length_size2+5) , 0 , 0)
                        target_models.add_entity(mg)
                        mg1 = mg.copy()
                        txt1 = txt0.copy()
                        mg1.translate(0 , -wd , 0)
                        mg1.dxf.color = 1
                        txt1.translate(0 , -wd , 0)
                        target_models.add_entity(mg1)
                        target_models.add_entity(txt1)
                        mg2=mg1.copy()
                        txt2=txt1.copy()
                        mg2.translate(0 , -wd+1 , 0)
                        txt2.translate(0 , -wd+1 , 0)
                        target_models.add_entity(mg2)
                        target_models.add_entity(txt2)
                        originy = Vec3(rfx + lg/2 - 2 , rfy + wd/2 - 0.25)
                        rcmm = target_models.add_lwpolyline(forms.translate(forms.box(4,0.5) , originy),
                                close = True , dxfattribs= {'color': 7})
                        rcmm.translate(-(2*length_size2+5) , -wd , 0)
                        txt00 = target_models.add_text(str(i+1), height = 0.3 , 
                               dxfattribs = {'style':'TXT','color':4}).set_placement((
                               rfx + lg/2 - 1.5 , rfy + wd/2 - 0.125),align = TextEntityAlignment.LEFT)
                        txt00.translate(-(2*length_size2+5) , -wd , 0)
                        r1 = rcmm.copy()
                        r1.dxf.color = 1
                        r1.translate(0,-2*wd + 2 ,0)
                        txt00.translate(0,-2*wd + 2,0)
                        target_models.add_entity(r1)
                        r2 = r1.copy()
                        txt11=txt00.copy()
                        r2.translate(4 , 0 , 0)
                        txt11.translate(4 , 0 , 0)
                        target_models.add_entity(r2)
                        target_models.add_entity(txt11)
                        r3 = r2.copy()
                        txt22=txt11.copy()
                        r3.translate(4 , 0 , 0)
                        txt22.translate(4 , 0 , 0)
                        target_models.add_entity(r3)
                        target_models.add_entity(txt22)
                else:
                    pass
            
            
                PL_s_x1 = reference_point_x_3
                PL_s_y1 = reference_point_y_3 + width_rec_next
                PL_e_x1 = reference_point_x_3 + length_rec_next
        
        
        
                line_1_horizontal = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_e_x1 , PL_s_y1) , 
                                       dxfattribs = {'color': 7} )
                line_3_horizontal = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_e_x1 , PL_s_y1) , 
                                       dxfattribs = {'color': free_side[1][3]} )
                line_1_vertical = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_s_x1 , PL_s_y1+3) , 
                                       dxfattribs = {'color': 7} )
                line_2_vertical = target_models.add_line( (PL_s_x1 + length_rec_next , PL_s_y1) ,
                                                 (PL_s_x1 + length_rec_next , PL_s_y1+3) , 
                                                 dxfattribs = {'color': 7})
        
        
                rect_3.translate(-(length_size2+5) , 0 , 0)
                line_3_horizontal.translate(-(length_size2 +5 ), 0 , 0)
        
        
        
                if not point_added:
                    pass
                else:
                    for ar in point_added:
                        p1 = ezdxf.math.Vec2(ar[0])
                        p2 = ezdxf.math.Vec2(ar[1])

                        arc = ConstructionArc.from_2p_radius(p1 , p2 , 0.2501 , True, True)
                        arc.add_to_layout(target_models , dxfattribs={'layer':'next object' , 'color':free_side[1][3]})

            else:
                pass
            
            for co_entity in p_first[5]:
                c_entity = co_entity.copy()
                c_entity.translate((length_size2)+5 , 0 , 0)
                c_entity.dxf.color = 7
                target_models.add_entity(c_entity)
            
            if free_side[1][8] == 0:
                target_models.add_text(str(substrate_name), height = 0.6 , 
                    dxfattribs = {'layer':'next object', 'style':'TXT','color':4}).set_placement((
                    length_rec_next/2 + reference_point_x_3,
                    reference_point_y_3+2) ,align = TextEntityAlignment.CENTER)
                target_models.add_text('code: '+str(code) , height = 0.6 ,
                    dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                    length_rec_next/2 + reference_point_x_3 , reference_point_y_3+1) , align = TextEntityAlignment.CENTER)
                target_models.add_text(str(i+1), height = 1, 
                    dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                    reference_point_x_3+1,reference_point_y_3+1) ,align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                    dxfattribs ={'layer':'next object','style':'TXT','color':4}).set_placement((
                    reference_point_x_3+1,reference_point_y_3+width_rec_next+1) , align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                    dxfattribs = {'layer':'first object','style':'TXT','color':4}).set_placement((
                    reference_point_x1 + 1 ,reference_point_y1+1) ,align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                    dxfattribs= {'layer':' first object','style':'TXT','color':4}).set_placement((
                    reference_point_x1 + length_size2 + 6 ,reference_point_y1+1) ,
                    align = TextEntityAlignment.LEFT)            
            
                for txt in p_next[6]:
                    x = (txt[0] - txt[1])/2 + txt[1]
                    y = txt[3] + 1
                    target_models.add_text(str(i+1) , height=1 , 
                         dxfattribs={'layer':'next object' , 'style':'TXT','color':4}).set_placement((
                         x , y), align = TextEntityAlignment.LEFT)
            
         
            else:
                target_models.add_text(str(substrate_name), height = 0.6 , 
                   dxfattribs = {'layer':'next object', 'style':'TXT','color':4}).set_placement((
                   3*length_size2/2 + reference_point_x1+5,
                   reference_point_y1+4) ,align = TextEntityAlignment.CENTER)
                target_models.add_text('code: '+str(code) , height = 0.6 ,
                   dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                   3*length_size2/2 + reference_point_x1+5 , reference_point_y1+3) , align = TextEntityAlignment.CENTER)
                #target_models.add_text(str(i+1), height = 1, 
                #        dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                #        reference_point_x_3+1,reference_point_y_3+1) ,align = TextEntityAlignment.LEFT)
                #target_models.add_text(str(i+1), height = 1, 
                #        dxfattribs ={'layer':'next object','style':'TXT','color':4}).set_placement((
                #        reference_point_x_3+1,reference_point_y_3+width_rec_next+1) , align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs = {'layer':'first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + 1 ,reference_point_y1+1) ,align = TextEntityAlignment.LEFT)
                target_models.add_text(str(i+1), height = 1, 
                        dxfattribs= {'layer':' first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + length_size2 + 6 ,reference_point_y1+1) ,
                        align = TextEntityAlignment.LEFT) 
                
                for txt in p_first[6]:
                    x = (txt[0] - txt[1])/2 + txt[1] + length_size2+5
                    y = txt[3] + 1
                    target_models.add_text(str(i+1) , height=1 , 
                                  dxfattribs={'layer':'next object' , 'style':'TXT','color':4}).set_placement((
                                  x , y), align = TextEntityAlignment.LEFT)
                
                
                
            for entity in target_models:
                if entity.dxf.layer == 'next object':
                    entity.translate(-(length_size2+5) ,  0 , 0 )
                    
                elif entity.dxf.layer == 'swing':
                    if free_side[1][8] == 0:
                        entity.translate(-(length_size2+5) ,  0 , 0 )
                    else:
                        pass          
        
            
            for en in target_models:
                if en.dxf.layer != 'cnc' and en.dxf.layer != 'nasb':
                    en.translate(-(reference_point_x1) + length_size2 + 5 + x6 ,  -(reference_point_y1) + distance_c, 0)
                else:
                    pass
                if en.dxf.layer == 'nasb':
                    pass
                else:
                    en.dxf.layer = 'cnc'
                    
                    
                    
            if free_side[1][8] == 0 :
                if free_side[1][4] == 1 :
                    if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                 close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                        
                            dd = dd - lg 
                        
                    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                               dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                               -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
                        dd = 0
                        for n in range(len(p_next[5])):
                            nc = p_next[5][n].copy()
                            xmx = max_abs(p_next[7][n])
                            xmn = min_abs(p_next[7][n])
                            ymx = max_abs(p_next[8][n])
                            ymn = min_abs(p_next[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 1
                            rfx = xmn - 1
                            rfy = ymn
                            nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                               dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                               -9 + dd + x6 , distance_c +0.5 + wd) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg

                    elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                  
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 2
                        rfx = xmn
                        rfy = ymn - 1
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                             -(lg/2 + 10) + x6 , distance_c + wd - 0.5) ,align = TextEntityAlignment.LEFT)
                   
                
                    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn - 1
                        rfy = ymn 
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    
                    elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_next[5])):
                            xmx = max_abs(p_next[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_next[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_next[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_next[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_next[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    else:
                        pass
                else:
                    pass
            else:
                if free_side[0][4] == 1 :
                    if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd + x6 - length_size2 , -ymn + distance_c + 1 , 0) #+ dd + x6
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                            
                            points = list(zip(p_first[7][n] , p_first[8][n] ))
                            min_x1 = min(points, key=lambda p: p[0])  
                            min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                            p1d = min(min_x_points1, key=lambda p: p[1])
                            p1d = (p1d[0] - xmn - 10 + dd + x6 - length_size2 , p1d[1]-ymn + distance_c+1)
                        
                            max_y11 = min(points, key=lambda p: p[1])  
                            max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                            p2d = min(max_y_points11, key=lambda p: p[0])
                            p2d = (p2d[0] - xmn - 10 + dd + x6 - length_size2 , p2d[1]-ymn + distance_c+1)
                                
                        
                            if round(p1d[0], 5) != round(p2d[0] , 5) and round(p1d[1] , 5) != round(p2d[1] , 5):
                                lg1 = abs(p2d[0] - p1d[0])
                                wd1 = abs(p2d[1] - p1d[1])
                                if wd1 >= 1 :
                                    ps1 = p1d
                                    pe1 = (p1d[0]-1, p1d[1])
                                    l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps2 = pe1
                                    pe2 = (pe1[0],pe1[1]-1)
                                    l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps3 = pe2
                                    pe3 = p2d
                                    l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                            
                                elif wd1 < 1 :
                                    df = (1 - wd1)/2
                                    ps1 = p1d
                                    pe1 = (p1d[0] , p1d[1]+df)
                                    l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps2 = pe1
                                    pe2 = (pe1[0] - 1, pe1[1])
                                    l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps3 = pe2
                                    pe3 = (pe2[0] , pe2[1] - 1)
                                    l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps4 = pe3
                                    pe4 = (p2d[0] , ps4[1])
                                    l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                    ps5 = pe4
                                    pe5 = p2d
                                    l5 = target_models.add_line( ps5 ,pe5 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                        
                            else:  
                                ps1 = (p1d[0] , p1d[1]+0.5)
                                pe1 = (ps1[0]-0.5 , ps1[1])
                                pl1 = [ps1,pe1]
                                l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                ps2 = pe1
                                pe2 = (pe1[0],pe1[1]-1)
                                l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                ps3 = pe2
                                pe3 = (ps3[0]+1 , ps3[1])
                                l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                ps4 = pe3
                                pe4 = (ps4[0] , ps4[1]+0.5)
                                l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                            
                            
                            
                            
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6 - length_size2, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -11 + lg/2 + dd + x6 - length_size2, distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                        
                            dd = dd - lg 
                        
                    elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==0:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn - 1
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6- length_size2 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                 close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6- length_size2, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                               dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                               -9 + dd + x6- length_size2 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==0:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 2
                            rfx = xmn
                            rfy = ymn - 1
                            nc.translate(-xmn - 10 + dd+x6- length_size2 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6- length_size2, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6- length_size2 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg
                        
                    elif free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==1:
                        dd = 0
                        for n in range(len(p_first[5])):
                            nc = p_first[5][n].copy()
                            xmx = max_abs(p_first[7][n])
                            xmn = min_abs(p_first[7][n])
                            ymx = max_abs(p_first[8][n])
                            ymn = min_abs(p_first[8][n])
                            lg = xmx - xmn + 2
                            wd = ymx - ymn + 1
                            rfx = xmn - 1
                            rfy = ymn
                            nc.translate(-xmn - 10 + dd+x6- length_size2 , -ymn + distance_c+1 , 0)
                            nc.dxf.layer = 'cnc'
                            nc.dxf.color = free_side[1][3]
                            target_models.add_entity(nc)
                        
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                               close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn - 10 + dd+x6- length_size2, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -9 + dd + x6 - length_size2, distance_c +0.5 + wd) ,align = TextEntityAlignment.LEFT)
                        
                            dd = dd - lg

                    elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==0:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                  
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 2
                        rfx = xmn
                        rfy = ymn - 1
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6- length_size2 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                     
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6- length_size2, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6- length_size2 , distance_c + wd - 0.5) ,align = TextEntityAlignment.LEFT)
                   
                
                    elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn - 1
                        rfy = ymn 
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6- length_size2 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6- length_size2, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6- length_size2 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn + 1
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6- length_size2 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6- length_size2, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6- length_size2 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    
                    elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==1:
                        x_mm = []
                        y_mm = []
                        for n in range(len(p_first[5])):
                            xmx = max_abs(p_first[7][n])
                            x_mm.append(xmx)
                            xmn = min_abs(p_first[7][n])
                            x_mm.append(xmn)
                            ymx = max_abs(p_first[8][n])
                            y_mm.append(ymx)
                            ymn = min_abs(p_first[8][n])
                            y_mm.append(ymn)
                           
                     
                        xmx = max_abs(x_mm)
                        xmn = min_abs(x_mm)
                        ymx = max_abs(y_mm)
                        ymn = min_abs(y_mm)
                    
                        lg = xmx - xmn 
                        wd = ymx - ymn + 1
                        rfx = xmn
                        rfy = ymn 
                    
                        for nv in p_first[5]:
                            nnv = nv.copy()
                            nnv.translate(-xmn - lg - 10 +x6- length_size2 , -ymn + distance_c+1 , 0)
                            nnv.dxf.layer = 'cnc'
                            nnv.dxf.color = free_side[1][3]
                            target_models.add_entity(nnv)
                        
                    
                    
                    
                        origin = Vec3(rfx,rfy)
                        rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                             close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                        rect_3dola.translate(-xmn -lg - 10 + x6- length_size2, -ymn+ distance_c+1,0)
                        text = target_models.add_text(str(i+1), height = 0.4, 
                             dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                            -(lg/2 + 10) + x6- length_size2 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    else:
                        pass
                    
                    
                    
        
            target_doc.saveas(tar_path)
            distance_c = distance_c + 45.37
    
    
        elif final_size == 'size3':
            length_size3 = 83
            width_size3 = 49.37
        
            x_min_swing = p_swing[3]
            y_min_swing = p_swing[4]
            x_max_swing = p_swing[3]+p_swing[0]
            y_max_swing = p_swing[4]+p_swing[1]
            #####

            reference_point_x11 = p_first[3] - (length_size3 - p_first[0])/2
            reference_point_y11 = p_first[4] - (width_size3 - p_first[1] )/2
            
            #####

            if reference_point_x11+12 < x_min_swing and reference_point_x11+71 > x_max_swing and reference_point_y11+10.69 < y_min_swing and reference_point_y11+46.69 > y_max_swing:
                reference_point_x1 = reference_point_x11
                reference_point_y1 = reference_point_y11
                
            
            elif reference_point_y11+10.69 < y_min_swing and reference_point_y11+46.69 > y_max_swing :
                reference_point_y1 = reference_point_y11
                
            
            elif reference_point_x11+12 < x_min_swing and reference_point_x11+71 > x_max_swing:
                reference_point_x1 = reference_point_x11
                
            
            
            if  (x_max_swing - x_min_swing) > 59 :
                c = 0
                window3 = tkinter.Tk()
                window3.withdraw()
                messagebox.showerror("Size Error", "I can not fix swing in bigest size(x direction).")
                
              
            
            elif reference_point_x11+12 > x_min_swing:
                diff_sx = reference_point_x11 + 12 - x_min_swing
                reference_point_x1 = reference_point_x11 - diff_sx - 1
                
            
            elif reference_point_x11 + 71 < x_max_swing :
                diff_sx = x_max_swing - (reference_point_x11+71)
                reference_point_x1 = reference_point_x11 + diff_sx + 1
                
            
        
            if (y_max_swing - y_min_swing) > 36:
                c = 0
                window3 = tkinter.Tk()
                window3.withdraw()
                messagebox.showerror("Size Error", "I can not fix swing in bigest size(y direction).")
                
        
        
        
            elif reference_point_y11 + 10.69 > y_min_swing:
                diff_sy = reference_point_y11+10.69 - y_min_swing
                reference_point_y1 = reference_point_y11 - diff_sy - 1
                

            elif reference_point_y11 + 46.69 < y_max_swing:
                diff_sy = y_max_swing - (reference_point_y11 + 46.69)
                reference_point_y1 = reference_point_y11 + diff_sy + 1
                
           
            if c == 1 :
                ########
                
                if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                    pass
                        
                elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==0:
                    dff = (reference_point_x1 + length_size3) - (p_first[3] + p_first[0]) 
                    reference_point_x1 = reference_point_x1 - dff
                    reference_point_y1 = reference_point_y1 
                    
                        
                elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==0:
                    dff = (reference_point_x1) - (p_first[3])
                    reference_point_x1 = reference_point_x1 - dff
                    reference_point_y1 = reference_point_y1
                   
                       
                elif free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==1:
                    dff = p_first[4] - reference_point_y1
                    reference_point_x1 = reference_point_x1
                    reference_point_y1 = reference_point_y1 + dff
                  
                        
                elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==0:
                    reference_point_x1 = reference_point_x1
                    reference_point_y1 = reference_point_y1

                    
                       
                elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==1:
                    dffy = p_first[4] - reference_point_y1
                    dffx = (reference_point_x1 + length_size3) - (p_first[3] + p_first[0])
                    reference_point_x1 = reference_point_x1 - dffx
                    reference_point_y1 = reference_point_y1 + dffy
                   
                    
                elif free_side[0][0]== 0 and free_side[0][1]==1 and free_side[0][2]==1:
                    dffy = p_first[4] - reference_point_y1
                    dffx = (reference_point_x1) - (p_first[3])
                    reference_point_x1 = reference_point_x1 - dffx
                    reference_point_y1 = reference_point_y1 + dffy
                   
                       
                else:
                    dff = p_first[4] - reference_point_y1
                    reference_point_x1 = reference_point_x1
                    reference_point_y1 = reference_point_y1 + dff
                
                ########
                c1 = (reference_point_x1+1.75 , reference_point_y1+8.15)
                c2 = (reference_point_x1+1.75 , reference_point_y1+27.5)
                c3 = (reference_point_x1+1.75 , reference_point_y1+47.15)
                c4 = (reference_point_x1+81.25 , reference_point_y1+47.15)
                c5 = (reference_point_x1+81.25 , reference_point_y1+27.5)
                c6 = (reference_point_x1+81.25 , reference_point_y1+8.15)
                c7 = (reference_point_x1+74.5 , reference_point_y1+1.5862)
                c8 = (reference_point_x1+41.5 , reference_point_y1+1.5862)
                c9 = (reference_point_x1+8.5 , reference_point_y1+1.5862)
        
        
                origin = Vec3(reference_point_x1,reference_point_y1)
                rect_1 = target_models.add_lwpolyline(forms.translate(forms.box(length_size3 , width_size3+3),
                                    origin), close = True , dxfattribs= {'color': free_side[0][3]})
                circle_11 = target_models.add_circle(c1 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_21 = target_models.add_circle(c2 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_31 = target_models.add_circle(c3 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_41 = target_models.add_circle(c4 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_51 = target_models.add_circle(c5 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_61 = target_models.add_circle(c6 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_71 = target_models.add_circle(c7 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_81 = target_models.add_circle(c8 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
                circle_91 = target_models.add_circle(c9 , radius=0.6 , dxfattribs={'color':free_side[0][3]})
        
                rect_0 = rect_1.copy()
                circle_10 = circle_11.copy()
                circle_20 = circle_21.copy()
                circle_30 = circle_31.copy()
                circle_40 = circle_41.copy()
                circle_50 = circle_51.copy()
                circle_60 = circle_61.copy()
                circle_70 = circle_71.copy()
                circle_80 = circle_81.copy()
                circle_90 = circle_91.copy()
        
                rect_0.translate(length_size3+5, 0, 0)
                rect_0.dxf.color = 5
                circle_10.translate(length_size3+5, 0, 0)
                circle_10.dxf.color = 5
                circle_20.translate(length_size3+5, 0, 0)
                circle_20.dxf.color = 5
                circle_30.translate(length_size3+5, 0, 0)
                circle_30.dxf.color = 5
                circle_40.translate(length_size3+5, 0, 0)
                circle_40.dxf.color = 5
                circle_50.translate(length_size3+5, 0, 0)
                circle_50.dxf.color = 5
                circle_60.translate(length_size3+5, 0, 0)
                circle_60.dxf.color = 5
                circle_70.translate(length_size3+5, 0, 0)
                circle_70.dxf.color = 5
                circle_80.translate(length_size3+5, 0, 0)
                circle_80.dxf.color = 5
                circle_90.translate(length_size3+5, 0, 0)
                circle_90.dxf.color = 5
        
        
    
                target_models.add_entity(rect_0)
                target_models.add_entity(circle_10)
                target_models.add_entity(circle_20)
                target_models.add_entity(circle_30)
                target_models.add_entity(circle_40)
                target_models.add_entity(circle_50)
                target_models.add_entity(circle_60)
                target_models.add_entity(circle_70)
                target_models.add_entity(circle_80)
                target_models.add_entity(circle_90)
            
                
                if free_side[1][8] == 1:
                    if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                        xy_pc = []
                        for idc in range(len(p_first[7])):
                            points = list(zip(p_first[7][idc], p_first[8][idc]))
                            min_x1 = min(points, key=lambda p: p[0])  
                            min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                            p1d = min(min_x_points1, key=lambda p: p[1])
                        
                            max_y11 = min(points, key=lambda p: p[1])  
                            max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                            p2d = min(max_y_points11, key=lambda p: p[0])
                        
                            if round(p1d[0], 5) != round(p2d[0] , 5) and round(p1d[1] , 5) != round(p2d[1] , 5):
                                lg = abs(p2d[0] - p1d[0])
                                wd = abs(p2d[1] - p1d[1])
                                if wd >= 1 :
                                    ps1 = p1d
                                    pe1 = (p1d[0]-1, p1d[1])
                                    l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                                
                                    ps2 = pe1
                                    pe2 = (pe1[0],pe1[1]-1)
                                    l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                                
                                    ps3 = pe2
                                    pe3 = p2d
                                    l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                            
                                elif wd < 1 :
                                    df = (1 - wd)/2
                                
                                    ps1 = p1d
                                    pe1 = (p1d[0] , p1d[1]+df)
                                    l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                                
                                    ps2 = pe1
                                    pe2 = (pe1[0] - 1, pe1[1])
                                    l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                                
                                    ps3 = pe2
                                    pe3 = (pe2[0] , pe2[1] - 1)
                                    l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                                
                                    ps4 = pe3
                                    pe4 = (p2d[0] , ps4[1])
                                    l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[0][3]} )
                                
                                    ps5 = pe4
                                    pe5 = p2d
                                    l5 = target_models.add_line( ps5 ,pe5 , dxfattribs = {'color': free_side[0][3]} )
                        
                            else:  
                                ps1 = (p1d[0] , p1d[1]+0.5)
                                pe1 = (ps1[0]-0.5 , ps1[1])
                                l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[0][3]} )
                            
                                ps2 = pe1
                                pe2 = (pe1[0],pe1[1]-1)
                                l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[0][3]} )
                            
                                ps3 = pe2
                                pe3 = (ps3[0]+1 , ps3[1])
                                l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[0][3]} )
                            
                                ps4 = pe3
                                pe4 = (ps4[0] , ps4[1]+0.5)
                                l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[0][3]} )
                     
                
                
                ###### add sight ######
                if free_side[1][8] == 0:
                    if free_side[1][7] == 1:
                        xy_pc = []
                        for iac in range(len(p_first[7])):
                            points = list(zip(p_first[7][iac], p_first[8][iac]))
                            min_x1 = min(points, key=lambda p: p[0])  
                            min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                            p1 = max(min_x_points1, key=lambda p: p[1]) 
    
                            max_x2 = max(points, key=lambda p: p[0])
                            max_x_points2 = [p for p in points if p[0] == max_x2[0]]
                            p2 = max(max_x_points2, key=lambda p: p[1])
    
                            max_y11 = max(points, key=lambda p: p[1])  
                            max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                            p11 = min(max_y_points11, key=lambda p: p[0])
    
                            max_y22 = max(points, key=lambda p: p[1])
                            max_y_points22 = [p for p in points if p[1] == max_y22[1]]
                            p22 = max(max_y_points22, key=lambda p: p[0])
    
                            if round(p1[0] , 5) != round(p11[0], 5) and round(p1[1] , 5) != round(p11[1], 5):
                                lg1 = p11[0] - p1[0]
                                wd1 = p11[1] - p1[1]
                                cp1 = (p11[0]-lg1/2 , p11[1]-wd1/2)
                            else:
                                cp1 = p11
        
                            if round(p2[0] , 5) != round(p22[0] , 5) and round(p2[1] , 5) != round(p22[1] , 5):
                                lg2 = p2[0] - p22[0]
                                wd2 = p22[1] - p2[1]
                                cp2 = (p22[0]+lg2/2 , p22[1]-wd2/2)
                            else:
                                cp2 = p22

                            l = []
                            l.append(cp1)
                            l.append(cp2)
                            xy_pc.append(l)
            
                        for cs in xy_pc :
                            circle_sight_L = target_models.add_circle(cs[0] , radius=2 , dxfattribs={'color':free_side[1][3]})
                            circle_sight_R = target_models.add_circle(cs[1] , radius=2 , dxfattribs={'color':free_side[1][3]})
                            circle_sight_L.translate(-(length_size3+ 5) , 0 ,0)
                            circle_sight_R.translate(-(length_size3+ 5) , 0 ,0)
                    else:
                        pass
                else:
                    pass
            
                for swing_rec in p_swing[6]:
                    lengths0 = abs(swing_rec[0] - swing_rec[1])+0.4
                    widths0 = abs(swing_rec[2] - swing_rec[3])+0.4
                    reference_point_x_s = swing_rec[1] - 0.2
                    reference_point_y_s = swing_rec[3] - 0.2
            
                    origin = Vec3(reference_point_x_s , reference_point_y_s)
                    rect_s = target_models.add_lwpolyline(forms.translate(forms.box(lengths0, widths0), origin), 
                                                 close = True , dxfattribs={'color':5})
                    rect_s.translate((length_size3)+5 , 0 , 0)
                if free_side[1][8] == 0:
                    #if p_next[0]+8 >= length_size3 :
                    #    length_rec_next = length_size3
                    #else:
                    length_rec_next = p_next[0] #+ 8
                    width_rec_next = p_next[1]
        
        
                    reference_point_x_3 = p_next[3] - (length_rec_next - p_next[0])/2
                    reference_point_y_3 = p_next[4] #- 4
        
                    the_diff = (reference_point_y1 + width_size3) - (reference_point_y_3 + 4 + width_rec_next)
                    width_rec_next = width_rec_next + the_diff
                
                    #if p_next[1] + 5 >= width_size3 :
                    #    width_rec_next = width_size3
                    #else:
                    #    width_rec_next = width_rec_next #+ 4
                    #######
                    H_f1 = 4
                    if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
                        length_rec_next = length_rec_next + 2*H_f1
                        width_rec_next = width_rec_next + 2*H_f1
                        reference_point_x_3 = reference_point_x_3 - H_f1
                        reference_point_y_3 = reference_point_y_3 - H_f1
                  
                        
                    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
                        length_rec_next = length_rec_next + H_f1
                        width_rec_next = width_rec_next + 2*H_f1
                        reference_point_x_3 = reference_point_x_3 - H_f1
                        reference_point_y_3 = reference_point_y_3 - H_f1
                    
                        
                    elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
                        length_rec_next = length_rec_next + H_f1
                        width_rec_next = width_rec_next + 2*H_f1
                        reference_point_x_3 = reference_point_x_3
                        reference_point_y_3 = reference_point_y_3 - H_f1
                   
                       
                    elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
                        length_rec_next = length_rec_next + 2* H_f1
                        width_rec_next = width_rec_next + H_f1
                        reference_point_x_3 = reference_point_x_3 - H_f1
                        reference_point_y_3 = reference_point_y_3
                  
                        
                    elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
                        length_rec_next = length_rec_next
                        width_rec_next = width_rec_next + 2*H_f1
                        reference_point_x_3 = reference_point_x_3
                        reference_point_y_3 = reference_point_y_3 - H_f1
                    
                       
                    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
                        length_rec_next = length_rec_next + H_f1
                        width_rec_next = width_rec_next + H_f1
                        reference_point_x_3 = reference_point_x_3 - H_f1
                        reference_point_y_3 = reference_point_y_3
                   
                    
                    elif free_side[1][0]== 0 and free_side[1][1]==1 and free_side[1][2]==1:
                        length_rec_next = length_rec_next + H_f1
                        width_rec_next = width_rec_next + H_f1
                        reference_point_x_3 = reference_point_x_3
                        reference_point_y_3 = reference_point_y_3
                   
                       
                    else:
                        length_rec_next = length_rec_next
                        width_rec_next = width_rec_next + H_f1
                        reference_point_x_3 = reference_point_x_3
                        reference_point_y_3 = reference_point_y_3
                    
        
        
        
                    origin_3 = Vec3(reference_point_x_3 , reference_point_y_3)
                    rect_3 = target_models.add_lwpolyline(forms.translate(forms.box(length_rec_next,width_rec_next+3),origin_3),
                                                   close=True , dxfattribs = {'color': free_side[1][3]})
        
                    #### add yaragh ####    
            
                    if free_side[1][5] == 1:
                        for yi in range(len(p_yg[7])):
                            xmx = max_abs(p_yg[7][yi])
                            xmn = min_abs(p_yg[7][yi])
                            ymx = max_abs(p_yg[8][yi])
                            ymn = min_abs(p_yg[8][yi])
                            lg = xmx - xmn + 4 
                            wd = ymx - ymn + 2.3 
                            rfx = xmn - 2
                            rfy = ymn - 0.3
                            originy = Vec3(rfx,rfy)
                            rcy = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd) , originy),
                                    close=True , dxfattribs = {'color': free_side[1][3], 'layer':'next object'})
                    else:
                        pass
                
                    #### add maghzi ####
            
                    if free_side[1][6] == 1:
                        for mi in range(len(p_mg[7])):
                            xmx = max_abs(p_mg[7][mi])
                            xmn = min_abs(p_mg[7][mi])
                            ymx = max_abs(p_mg[8][mi])
                            ymn = min_abs(p_mg[8][mi])
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 1
                            rfx = xmn - 0.5
                            rfy = ymn - 0.5
                            originy = Vec3(rfx,rfy)
                            rcm = target_models.add_lwpolyline(forms.translate(forms.box(lg,wd) , originy),
                                close = True , dxfattribs= {'color': 3})
                            txt0=target_models.add_text(str(i+1), height = 0.3 , 
                                dxfattribs = {'style':'TXT','color':4}).set_placement((
                                rfx+0.625,rfy+0.625),align = TextEntityAlignment.LEFT)
                    
                            rcm.translate(-(2*length_size3+5) , 0 , 0)
                            txt0.translate(-(2*length_size3+5) , 0 , 0)
                            mg = p_mg[5][mi].copy()
                            mg.dxf.color = 7
                            mg.translate(-(2*length_size3+5) , 0 , 0)
                            target_models.add_entity(mg)
                            mg1 = mg.copy()
                            txt1 = txt0.copy()
                            mg1.translate(0 , -wd , 0)
                            mg1.dxf.color = 1
                            txt1.translate(0 , -wd , 0)
                            target_models.add_entity(mg1)
                            target_models.add_entity(txt1)
                            mg2=mg1.copy()
                            txt2=txt1.copy()
                            mg2.translate(0 , -wd+1 , 0)
                            txt2.translate(0 , -wd+1 , 0)
                            target_models.add_entity(mg2)
                            target_models.add_entity(txt2)
                            originy = Vec3(rfx + lg/2 - 2 , rfy + wd/2 - 0.25)
                            rcmm = target_models.add_lwpolyline(forms.translate(forms.box(4,0.5) , originy),
                                close = True , dxfattribs= {'color': 7})
                            rcmm.translate(-(2*length_size3+5) , -wd , 0)
                            txt00 = target_models.add_text(str(i+1), height = 0.3 , 
                               dxfattribs = {'style':'TXT','color':4}).set_placement((
                               rfx + lg/2 - 1.5 , rfy + wd/2 - 0.125),align = TextEntityAlignment.LEFT)
                            txt00.translate(-(2*length_size3+5) , -wd , 0)
                            r1 = rcmm.copy()
                            r1.dxf.color = 1
                            r1.translate(0,-2*wd + 2 ,0)
                            txt00.translate(0,-2*wd + 2,0)
                            target_models.add_entity(r1)
                            r2 = r1.copy()
                            txt11=txt00.copy()
                            r2.translate(4 , 0 , 0)
                            txt11.translate(4 , 0 , 0)
                            target_models.add_entity(r2)
                            target_models.add_entity(txt11)
                            r3 = r2.copy()
                            txt22=txt11.copy()
                            r3.translate(4 , 0 , 0)
                            txt22.translate(4 , 0 , 0)
                            target_models.add_entity(r3)
                            target_models.add_entity(txt22)
                    
                    else: 
                        pass
            
            
                
                
                
                    PL_s_x1 = reference_point_x_3
                    PL_s_y1 = reference_point_y_3 + width_rec_next
                    PL_e_x1 = reference_point_x_3 + length_rec_next
        
        
        
                    line_1_horizontal = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_e_x1 , PL_s_y1) , 
                                       dxfattribs = {'color': 7} )
                    line_3_horizontal = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_e_x1 , PL_s_y1) , 
                                       dxfattribs = {'color': free_side[1][3]} )
                    line_1_vertical = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_s_x1 , PL_s_y1+3) , 
                                       dxfattribs = {'color': 7} )
                    line_2_vertical = target_models.add_line( (PL_s_x1 + length_rec_next , PL_s_y1) ,
                                                 (PL_s_x1 + length_rec_next , PL_s_y1+3) , 
                                                 dxfattribs = {'color': 7})
        
                    rect_3.translate(-(length_size3+5) , 0 , 0)
                    line_3_horizontal.translate(-(length_size3 +5 ), 0 , 0)
        
        
                    if not point_added:
                        pass
                    else:
                        for ar in point_added:
                            p1 = ezdxf.math.Vec2(ar[0])
                            p2 = ezdxf.math.Vec2(ar[1])
            
                            arc = ConstructionArc.from_2p_radius(p1 , p2 , 0.2501 , True, True)
                            arc.add_to_layout(target_models , dxfattribs={'layer':'next object' , 'color':free_side[1][3]})

                else:
                    pass
                
                for co_entity in p_first[5]:
                    c_entity = co_entity.copy()
                    c_entity.translate((length_size3)+5 , 0 , 0)
                    c_entity.dxf.color = 7
                    target_models.add_entity(c_entity)
            
                if free_side[1][8] == 0:
                    target_models.add_text(str(substrate_name), height = 0.6 , 
                       dxfattribs = {'layer':'next object', 'style':'TXT','color':4}).set_placement((
                       length_rec_next/2 + reference_point_x_3,
                       reference_point_y_3+2) ,align = TextEntityAlignment.CENTER)
                    target_models.add_text('code: '+str(code) , height = 0.6 ,
                       dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                       length_rec_next/2 + reference_point_x_3 , reference_point_y_3+1) , align = TextEntityAlignment.CENTER)
                    target_models.add_text(str(i+1), height = 1, 
                       dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                       reference_point_x_3+1,reference_point_y_3+1) ,align = TextEntityAlignment.LEFT)
                    target_models.add_text(str(i+1), height = 1, 
                       dxfattribs ={'layer':'next object','style':'TXT','color':4}).set_placement((
                       reference_point_x_3+1,reference_point_y_3+width_rec_next+1) , align = TextEntityAlignment.LEFT)
                    target_models.add_text(str(i+1), height = 1, 
                       dxfattribs = {'layer':'first object','style':'TXT','color':4}).set_placement((
                       reference_point_x1 + 1 ,reference_point_y1+1) ,align = TextEntityAlignment.LEFT)
                    target_models.add_text(str(i+1), height = 1, 
                       dxfattribs= {'layer':' first object','style':'TXT','color':4}).set_placement((
                       reference_point_x1 + length_size3 + 6 ,reference_point_y1+1) ,
                       align = TextEntityAlignment.LEFT)            
            
                    for txt in p_next[6]:
                        x = (txt[0] - txt[1])/2 + txt[1]
                        y = txt[3] + 1
                        target_models.add_text(str(i+1) , height=1 , 
                                  dxfattribs={'layer':'next object' , 'style':'TXT','color':4}).set_placement((
                                  x , y), align = TextEntityAlignment.LEFT)
                else:
                    target_models.add_text(str(substrate_name), height = 0.6 , 
                       dxfattribs = {'layer':'next object', 'style':'TXT','color':4}).set_placement((
                       3*length_size3/2 + reference_point_x1+5,
                       reference_point_y1+4) ,align = TextEntityAlignment.CENTER)
                    target_models.add_text('code: '+str(code) , height = 0.6 ,
                       dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                       3*length_size3/2 + reference_point_x1+5 , reference_point_y1+3) , align = TextEntityAlignment.CENTER)
                    #target_models.add_text(str(i+1), height = 1, 
                    #        dxfattribs = {'layer':'next object','style':'TXT','color':4}).set_placement((
                    #        reference_point_x_3+1,reference_point_y_3+1) ,align = TextEntityAlignment.LEFT)
                    #target_models.add_text(str(i+1), height = 1, 
                    #        dxfattribs ={'layer':'next object','style':'TXT','color':4}).set_placement((
                    #        reference_point_x_3+1,reference_point_y_3+width_rec_next+1) , align = TextEntityAlignment.LEFT)
                    target_models.add_text(str(i+1), height = 1, 
                        dxfattribs = {'layer':'first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + 1 ,reference_point_y1+1) ,align = TextEntityAlignment.LEFT)
                    target_models.add_text(str(i+1), height = 1, 
                        dxfattribs= {'layer':' first object','style':'TXT','color':4}).set_placement((
                        reference_point_x1 + length_size3 + 6 ,reference_point_y1+1) ,
                        align = TextEntityAlignment.LEFT) 
                
                    for txt in p_first[6]:
                        x = (txt[0] - txt[1])/2 + txt[1] + length_size3+5
                        y = txt[3] + 1
                        target_models.add_text(str(i+1) , height=1 , 
                                  dxfattribs={'layer':'next object' , 'style':'TXT','color':4}).set_placement((
                                  x , y), align = TextEntityAlignment.LEFT)
                    
            
            
            
                for entity in target_models:
                    if entity.dxf.layer == 'next object':
                        entity.translate(-(length_size3+5) ,  0 , 0 )
                    
                    elif entity.dxf.layer == 'swing':
                        if free_side[1][8] == 0:
                            entity.translate(-(length_size3+5) ,  0 , 0 )
                        else:
                            pass 
                    else:
                        pass
            
            
                for en in target_models:
                    if en.dxf.layer != 'cnc' and en.dxf.layer != 'nasb': 
                        en.translate(-(reference_point_x1) + length_size3 + 5 + x6 ,  -(reference_point_y1) +  distance_c, 0)
                    else:
                        pass
                    if en.dxf.layer == 'nasb':
                        pass
                    else:
                        en.dxf.layer = 'cnc'

                 
                if free_side[1][8] == 0:
                    if free_side[1][4] == 1:
                        if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
                            dd = 0
                            for n in range(len(p_next[5])):
                                nc = p_next[5][n].copy()
                                xmx = max_abs(p_next[7][n])
                                xmn = min_abs(p_next[7][n])
                                ymx = max_abs(p_next[8][n])
                                ymn = min_abs(p_next[8][n])
                                lg = xmx - xmn + 2
                                wd = ymx - ymn + 2
                                rfx = xmn - 1
                                rfy = ymn - 1
                                nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg 
                        
                        elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
                            dd = 0
                            for n in range(len(p_next[5])):
                                nc = p_next[5][n].copy()
                                xmx = max_abs(p_next[7][n])
                                xmn = min_abs(p_next[7][n])
                                ymx = max_abs(p_next[8][n])
                                ymn = min_abs(p_next[8][n])
                                lg = xmx - xmn + 1
                                wd = ymx - ymn + 2
                                rfx = xmn - 1
                                rfy = ymn - 1
                                nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg
                        
                        elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
                            dd = 0
                            for n in range(len(p_next[5])):
                                nc = p_next[5][n].copy()
                                xmx = max_abs(p_next[7][n])
                                xmn = min_abs(p_next[7][n])
                                ymx = max_abs(p_next[8][n])
                                ymn = min_abs(p_next[8][n])
                                lg = xmx - xmn + 1
                                wd = ymx - ymn + 2
                                rfx = xmn
                                rfy = ymn - 1
                                nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -9 + dd + x6 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg
                        
                        elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
                            dd = 0
                            for n in range(len(p_next[5])):
                                nc = p_next[5][n].copy()
                                xmx = max_abs(p_next[7][n])
                                xmn = min_abs(p_next[7][n])
                                ymx = max_abs(p_next[8][n])
                                ymn = min_abs(p_next[8][n])
                                lg = xmx - xmn + 2
                                wd = ymx - ymn + 1
                                rfx = xmn - 1
                                rfy = ymn
                                nc.translate(-xmn - 10 + dd+x6 , -ymn + distance_c+1 , 0)
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -9 + dd + x6 , distance_c +0.5 + wd) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg

                        elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_next[5])):
                                xmx = max_abs(p_next[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_next[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_next[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_next[8][n])
                                y_mm.append(ymn)
                  
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn 
                            wd = ymx - ymn + 2
                            rfx = xmn
                            rfy = ymn - 1
                    
                            for nv in p_next[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                        
                    
                    
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                 close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6 , distance_c + wd - 0.5) ,align = TextEntityAlignment.LEFT)
                   
                
                        elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_next[5])):
                                xmx = max_abs(p_next[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_next[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_next[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_next[8][n])
                                y_mm.append(ymn)
                           
                     
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 1
                            rfx = xmn - 1
                            rfy = ymn 
                    
                            for nv in p_next[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                        
                    
                    
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                        elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==1:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_next[5])):
                                xmx = max_abs(p_next[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_next[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_next[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_next[8][n])
                                y_mm.append(ymn)
                           
                     
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 1
                            rfx = xmn
                            rfy = ymn 
                    
                            for nv in p_next[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                        
                    
                    
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    
                        elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==1:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_next[5])):
                                xmx = max_abs(p_next[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_next[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_next[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_next[8][n])
                                y_mm.append(ymn)
                           
                     
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn 
                            wd = ymx - ymn + 1
                            rfx = xmn
                            rfy = ymn 
                    
                            for nv in p_next[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                        
                    
                    
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                        else:
                            pass
                    else:
                        pass
                else:
                    if free_side[0][4] == 1 :
                        if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
                            dd = 0
                            for n in range(len(p_first[5])):
                                nc = p_first[5][n].copy()
                                xmx = max_abs(p_first[7][n])
                                xmn = min_abs(p_first[7][n])
                                ymx = max_abs(p_first[8][n])
                                ymn = min_abs(p_first[8][n])
                                lg = xmx - xmn + 2
                                wd = ymx - ymn + 2
                                rfx = xmn - 1
                                rfy = ymn - 1
                                nc.translate(-xmn - 10 + dd + x6 - length_size3 , -ymn + distance_c + 1 , 0) #+ dd + x6
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                            
                                points = list(zip(p_first[7][n] , p_first[8][n] ))
                                min_x1 = min(points, key=lambda p: p[0])  
                                min_x_points1 = [p for p in points if p[0] == min_x1[0]]  
                                p1d = min(min_x_points1, key=lambda p: p[1])
                                p1d = (p1d[0] - xmn - 10 + dd + x6 - length_size3 , p1d[1]-ymn + distance_c+1)
                        
                                max_y11 = min(points, key=lambda p: p[1])  
                                max_y_points11 = [p for p in points if p[1] == max_y11[1]]  
                                p2d = min(max_y_points11, key=lambda p: p[0])
                                p2d = (p2d[0] - xmn - 10 + dd + x6 - length_size3 , p2d[1]-ymn + distance_c+1)
                                
                        
                                if round(p1d[0], 5) != round(p2d[0] , 5) and round(p1d[1] , 5) != round(p2d[1] , 5):
                                    lg1 = abs(p2d[0] - p1d[0])
                                    wd1 = abs(p2d[1] - p1d[1])
                                    if wd1 >= 1 :
                                        ps1 = p1d
                                        pe1 = (p1d[0]-1, p1d[1])
                                        l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                        ps2 = pe1
                                        pe2 = (pe1[0],pe1[1]-1)
                                        l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                        ps3 = pe2
                                        pe3 = p2d
                                        l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                            
                                    elif wd1 < 1 :
                                        df = (1 - wd1)/2
                                        ps1 = p1d
                                        pe1 = (p1d[0] , p1d[1]+df)
                                        l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                        ps2 = pe1
                                        pe2 = (pe1[0] - 1, pe1[1])
                                        l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                        ps3 = pe2
                                        pe3 = (pe2[0] , pe2[1] - 1)
                                        l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                        ps4 = pe3
                                        pe4 = (p2d[0] , ps4[1])
                                        l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                                
                                        ps5 = pe4
                                        pe5 = p2d
                                        l5 = target_models.add_line( ps5 ,pe5 , dxfattribs = {'color': free_side[1][3]
                                                                                         , 'layer':'cnc'} )
                        
                                else:  
                                    ps1 = (p1d[0] , p1d[1]+0.5)
                                    pe1 = (ps1[0]-0.5 , ps1[1])
                                    pl1 = [ps1,pe1]
                                    l1 = target_models.add_line( ps1 ,pe1 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                    ps2 = pe1
                                    pe2 = (pe1[0],pe1[1]-1)
                                    l2 = target_models.add_line( ps2 ,pe2 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                    ps3 = pe2
                                    pe3 = (ps3[0]+1 , ps3[1])
                                    l3 = target_models.add_line( ps3 ,pe3 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                                    ps4 = pe3
                                    pe4 = (ps4[0] , ps4[1]+0.5)
                                    l4 = target_models.add_line( ps4 ,pe4 , dxfattribs = {'color': free_side[1][3]
                                                                                     , 'layer':'cnc'} )
                            
                            
                            
                            
                            
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6 - length_size3, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -11 + lg/2 + dd + x6 - length_size3, distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg 
                        
                        elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==0:
                            dd = 0
                            for n in range(len(p_first[5])):
                                nc = p_first[5][n].copy()
                                xmx = max_abs(p_first[7][n])
                                xmn = min_abs(p_first[7][n])
                                ymx = max_abs(p_first[8][n])
                                ymn = min_abs(p_first[8][n])
                                lg = xmx - xmn + 1
                                wd = ymx - ymn + 2
                                rfx = xmn - 1
                                rfy = ymn - 1
                                nc.translate(-xmn - 10 + dd+x6- length_size3 , -ymn + distance_c+1 , 0)
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6- length_size3, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -9 + dd + x6- length_size3 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg
                        
                        elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==0:
                            dd = 0
                            for n in range(len(p_first[5])):
                                nc = p_first[5][n].copy()
                                xmx = max_abs(p_first[7][n])
                                xmn = min_abs(p_first[7][n])
                                ymx = max_abs(p_first[8][n])
                                ymn = min_abs(p_first[8][n])
                                lg = xmx - xmn + 1
                                wd = ymx - ymn + 2
                                rfx = xmn
                                rfy = ymn - 1
                                nc.translate(-xmn - 10 + dd+x6- length_size3 , -ymn + distance_c+1 , 0)
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6- length_size3, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -9 + dd + x6- length_size3 , distance_c +0.5) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg
                        
                        elif free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==1:
                            dd = 0
                            for n in range(len(p_first[5])):
                                nc = p_first[5][n].copy()
                                xmx = max_abs(p_first[7][n])
                                xmn = min_abs(p_first[7][n])
                                ymx = max_abs(p_first[8][n])
                                ymn = min_abs(p_first[8][n])
                                lg = xmx - xmn + 2
                                wd = ymx - ymn + 1
                                rfx = xmn - 1
                                rfy = ymn
                                nc.translate(-xmn - 10 + dd+x6- length_size3 , -ymn + distance_c+1 , 0)
                                nc.dxf.layer = 'cnc'
                                nc.dxf.color = free_side[1][3]
                                target_models.add_entity(nc)
                        
                                origin = Vec3(rfx,rfy)
                                rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                    close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                                rect_3dola.translate(-xmn - 10 + dd+x6- length_size3, -ymn+ distance_c+1,0)
                                text = target_models.add_text(str(i+1), height = 0.4, 
                                    dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                    -9 + dd + x6 - length_size3, distance_c +0.5 + wd) ,align = TextEntityAlignment.LEFT)
                                dd = dd - lg

                        elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==0:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_first[5])):
                                xmx = max_abs(p_first[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_first[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_first[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_first[8][n])
                                y_mm.append(ymn)
                  
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn 
                            wd = ymx - ymn + 2
                            rfx = xmn
                            rfy = ymn - 1
                    
                            for nv in p_first[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6- length_size3 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6- length_size3, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6- length_size3 , distance_c + wd - 0.5) ,align = TextEntityAlignment.LEFT)
                   
                
                        elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==1:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_first[5])):
                                xmx = max_abs(p_first[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_first[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_first[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_first[8][n])
                                y_mm.append(ymn)
                           
                     
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 1
                            rfx = xmn - 1
                            rfy = ymn 
                    
                            for nv in p_first[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6- length_size3 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                        
                    
                    
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6- length_size3, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6- length_size3 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                        elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==1:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_first[5])):
                                xmx = max_abs(p_first[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_first[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_first[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_first[8][n])
                                y_mm.append(ymn)
                           
                     
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn + 1
                            wd = ymx - ymn + 1
                            rfx = xmn
                            rfy = ymn 
                    
                            for nv in p_first[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6- length_size3 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                        
                    
                    
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6- length_size3, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                 dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6- length_size3 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                    
                    
                        elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==1:
                            x_mm = []
                            y_mm = []
                            for n in range(len(p_first[5])):
                                xmx = max_abs(p_first[7][n])
                                x_mm.append(xmx)
                                xmn = min_abs(p_first[7][n])
                                x_mm.append(xmn)
                                ymx = max_abs(p_first[8][n])
                                y_mm.append(ymx)
                                ymn = min_abs(p_first[8][n])
                                y_mm.append(ymn)
                           
                     
                            xmx = max_abs(x_mm)
                            xmn = min_abs(x_mm)
                            ymx = max_abs(y_mm)
                            ymn = min_abs(y_mm)
                    
                            lg = xmx - xmn 
                            wd = ymx - ymn + 1
                            rfx = xmn
                            rfy = ymn 
                    
                            for nv in p_first[5]:
                                nnv = nv.copy()
                                nnv.translate(-xmn - lg - 10 +x6- length_size3 , -ymn + distance_c+1 , 0)
                                nnv.dxf.layer = 'cnc'
                                nnv.dxf.color = free_side[1][3]
                                target_models.add_entity(nnv)
                        
                    
                    
                    
                            origin = Vec3(rfx,rfy)
                            rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(lg , wd),origin), 
                                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'cnc'})
                            rect_3dola.translate(-xmn -lg - 10 + x6- length_size3, -ymn+ distance_c+1,0)
                            text = target_models.add_text(str(i+1), height = 0.4, 
                                dxfattribs = {'layer':'cnc','style':'TXT','color':4}).set_placement((
                                -(lg/2 + 10) + x6- length_size3 , distance_c + wd + 0.5) ,align = TextEntityAlignment.LEFT)
                        else:
                            pass

                distance_c = distance_c + 61.37
        
        
        
            
            else:
                for entity in target_models:
                    if entity.dxf.layer == 'swing' or entity.dxf.layer == 'first object' or entity.dxf.layer == 'next object':
                        entity.destroy()
        
        
            target_doc.saveas(tar_path) 
            
        return distance_c
    
    
    
    cc = create_the_substrates(distance_c, i , substrate_name , code , source_file_path  , free_side , x6)
    os.remove(test_path)
    if c == 1:
        update_i_distance_cnc( cc , i , substrate_name , x6)
    else:
        pass
    
    
    
def create_the_nasb_sub(distance_y, i , substrate_name , code , source_file_path  , free_side , x6):
    get_recently_started_tasks()
    directory_path = 'C:/Users/A.Mohammadlou/Desktop/software_test/'

    if os.path.exists(directory_path):
        for filename in os.listdir(directory_path):
            if filename == 'test.dxf':
                file_path = os.path.join(directory_path, filename)
                os.remove(file_path)
    
    target_doc = ezdxf.readfile(tar_path)
    target_models = target_doc.modelspace()
    for entityd in target_models:
        if (entityd.dxf.layer != 'nasb' and entityd.dxf.layer != 'cnc' 
            and entityd.dxf.layer != 'daghi' and entityd.dxf.layer != 'kalaf' 
            and entityd.dxf.layer != 'Defpoints' and entityd.dxf.layer != 'fesharki'
           and entityd.dxf.layer != 'zip' and entityd.dxf.layer != 'tighe zani'):
            entityd.destroy()
        else:
            pass
    
    acad = win32com.client.Dispatch("AutoCAD.Application") 
    acad.Documents.Open(source_file_path)
    time.sleep(1)
    doc = acad.ActiveDocument
    
    new_layer1 = doc.Layers.Add('maghar1')
    new_layer2 = doc.Layers.Add('maghar2')
    
    time.sleep(2)
    
    selSet1 = acad.ActiveDocument.SelectionSets.Add('01')
    selSet1.SelectOnScreen()
    L01 = 0
    for obj in selSet1:
        obj.layer = 'maghar1'
        L011 = obj.Length
        L01 += L011

    
    time.sleep(1)
    selSet2 = acad.ActiveDocument.SelectionSets.Add('02')
    selSet2.SelectOnScreen()
    for obj in selSet2:
        obj.layer = 'maghar2'
        L03 = obj.Length
    
    
    time.sleep(1)
    doc.Saveas("C:/Users/A.Mohammadlou/Desktop/software_test/test.dxf")
    time.sleep(1)
    doc.Close()
    
    source_doc = ezdxf.readfile("C:/Users/A.Mohammadlou/Desktop/software_test/test.dxf")
    
    source_models = source_doc.modelspace()
    

    
    for entity in source_models:
        if entity.dxf.layer == 'maghar1':
            entity_copy1 = entity.copy()
            if free_side[1][5] == '':
                pass
            else:
                teta = (int(free_side[1][5])/180) * math.pi
                entity_copy1.rotate_z(teta)
                
            if str(free_side[1][6]) == 'x':
                entity_copy1.transform(Matrix44.scale(1,-1,1))
            elif str(free_side[1][6]) == 'y':
                entity_copy1.transform(Matrix44.scale(-1,1,1))
            else:
                pass
                
            upright(entity_copy1)
            target_models.add_entity(entity_copy1)
            target_doc.saveas(tar_path)   
    
    the_first_object = []
    x_max_min_f=[]
    y_max_min_f=[]
    L11f = 0
    for entity in target_models:
        if entity.dxf.layer == 'maghar1':
            the_first_object.append(entity)
            polyline_points = entity.get_points()
            x_max_min=[]
            y_max_min=[]
            for j in range (0,len(polyline_points)):
                x_max_min.append(polyline_points[j][0])
                x_max_min_f.append(polyline_points[j][0])
            for j in range (0,len(polyline_points)):
                y_max_min.append(polyline_points[j][1])
                y_max_min_f.append(polyline_points[j][1])        
            L11 = line_length(x_max_min , y_max_min)
            L11f += L11
    
    
    
    x_max1=max_abs(x_max_min_f)
    x_min1=min_abs(x_max_min_f)
    y_max1=max_abs(y_max_min_f)
    y_min1=min_abs(y_max_min_f)
    Length1 = x_max1 - x_min1
    Width1 = y_max1 - y_min1

    if L01 > L11f :
        H_i= abs((L01 -  L11f)/2)
        H_f1 = H_i + 5
    else:
        H_f1 = 5
    if free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==0:
        final_Length1 = Length1 + 2*H_f1
        final_Width1 = Width1 + 2*H_f1
        reference_point_x1 = x_min1 - H_f1
        reference_point_y1 = y_min1 - H_f1
    elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==0:
        final_Length1 = Length1 + H_f1
        final_Width1 = Width1 + 2*H_f1
        reference_point_x1 = x_min1 - H_f1
        reference_point_y1 = y_min1 - H_f1
    elif free_side[0][0]==0 and free_side[0][1]==1 and free_side[0][2]==0:
        final_Length1 = Length1 + H_f1
        final_Width1 = Width1 + 2*H_f1
        reference_point_x1 = x_min1
        reference_point_y1 = y_min1 - H_f1
    elif free_side[0][0]==0 and free_side[0][1]==0 and free_side[0][2]==1:
        final_Length1 = Length1 + 2*H_f1
        final_Width1 = Width1 + H_f1
        reference_point_x1 = x_min1- H_f1
        reference_point_y1 = y_min1
    elif free_side[0][0]==1 and free_side[0][1]==1 and free_side[0][2]==0:
        final_Length1 = Length1
        final_Width1 = Width1 + 2*H_f1
        reference_point_x1 = x_min1
        reference_point_y1 = y_min1 - H_f1
    elif free_side[0][0]==1 and free_side[0][1]==0 and free_side[0][2]==1:
        final_Length1 = Length1 + H_f1
        final_Width1 = Width1 + H_f1
        reference_point_x1 = x_min1 - H_f1
        reference_point_y1 = y_min1
    elif free_side[0][0]== 0 and free_side[0][1]==1 and free_side[0][2]==1:
        final_Length1 = Length1 + H_f1
        final_Width1 = Width1 + H_f1
        reference_point_x1 = x_min1
        reference_point_y1 = y_min1
    else :
        final_Length1 = Length1 
        final_Width1 = Width1 + H_f1
        reference_point_x1 = x_min1
        reference_point_y1 = y_min1
                
                 

                
    origin = Vec3(reference_point_x1,reference_point_y1)
    rect_1 = target_models.add_lwpolyline(forms.translate(forms.box(final_Length1,final_Width1+3),
                                    origin), close = True , dxfattribs= {'color': free_side[0][3]})
    rect1_points = rect_1.get_points()
    x_max_min_rect1=[]
    y_max_min_rect1=[]
    for j in range (0,len(rect1_points)):
        x_max_min_rect1.append(rect1_points[j][0])
    for j in range (0,len(rect1_points)):
        y_max_min_rect1.append(rect1_points[j][1])
               
    x_max_rect1=max_abs(x_max_min_rect1)
    x_min_rect1=min_abs(x_max_min_rect1)
    y_max_rect1=max_abs(y_max_min_rect1)
    y_min_rect1=min_abs(y_max_min_rect1)
    trans_rect1_x = - reference_point_x1 + final_Length1 + 5
    trans_rect1_y = - reference_point_y1
    trans_first_object_x = trans_rect1_x + final_Length1 + 5
    trans_first_object_y = trans_rect1_y 
    rect_1c = rect_1.copy()
    rect_1.translate(trans_rect1_x + x6,trans_rect1_y+distance_y,0)
    rect_1.dxf.color = free_side[0][3]
    rect_1.dxf.layer = 'nasb'
    rect_1c.translate(trans_first_object_x + x6 , trans_first_object_y + distance_y , 0)
    rect_1c.dxf.color = 5
    rect_1c.dxf.layer = 'nasb'
    target_models.add_entity(rect_1c)
    for ent in target_models:
        if ent.dxf.layer == 'maghar1':
            encopy = ent.copy()
            ent.translate(trans_first_object_x + x6, trans_first_object_y + distance_y , 0 )
            ent.dxf.color = 7
            ent.dxf.layer = 'nasb'
            encopy.translate(trans_first_object_x - (final_Length1 + 5) + x6 , 
                                 trans_first_object_y + distance_y , 0 )
                
            encopy.dxf.color = free_side[0][3]
            encopy.dxf.layer = 'nasb'
            target_models.add_entity(encopy)
        else:
            pass

    target_doc.saveas(tar_path)




    the_next_object = []
    x_max_min_f=[]
    y_max_min_f=[]
    L33f = 0
    for entity in source_models:
        if entity.dxf.layer == 'maghar2':
            entity_copy1 = entity.copy()
            entity_copy2 = entity.copy()
            if free_side[1][5] == '':
                pass
            else:
                teta = (int(free_side[1][5])/180) * math.pi
                entity_copy1.rotate_z(teta)
                entity_copy2.rotate_z(teta)
                
            if str(free_side[1][6]) == 'x':
                entity_copy1.transform(Matrix44.scale(1,-1,1))
                entity_copy2.transform(Matrix44.scale(1,-1,1))
            elif str(free_side[1][6]) == 'y':
                entity_copy1.transform(Matrix44.scale(-1,1,1))
                entity_copy2.transform(Matrix44.scale(-1,1,1))
            else:
                pass

            upright(entity_copy1)
            upright(entity_copy2)
            target_models.add_entity(entity_copy1)
            target_doc.saveas(tar_path)   
                
    for entity in target_models:
        if entity.dxf.layer == 'maghar2':
            the_next_object.append(entity)
            dola = entity.copy()
            polyline_points = entity.get_points()
            x_max_min=[]
            y_max_min=[]
            for j in range (0,len(polyline_points)):
                x_max_min.append(polyline_points[j][0])
                x_max_min_f.append(polyline_points[j][0])
            for j in range (0,len(polyline_points)):
                y_max_min.append(polyline_points[j][1])
                y_max_min_f.append(polyline_points[j][1])
            L13 = line_length(x_max_min , y_max_min)
            L33f += L13
        

    x_max3=max_abs(x_max_min_f)
    x_min3=min_abs(x_max_min_f)
    y_max3=max_abs(y_max_min_f)
    y_min3=min_abs(y_max_min_f)
    Length3 = x_max3 - x_min3
    Width3 = y_max3 - y_min3
    
    
                    
    if L03 > L33f :
        H_i= abs((L03 -  L13)/2)
        H_f1 = H_i + 5
    else:
        H_i = 0
        H_f1 = 5
    final_Length3 = Length3 + 2*H_f1
    final_Width3 = Width3 + 2*H_f1
    reference_point_x3 = x_min3 - H_f1
    reference_point_y3 = y_min3 - H_f1                 
    if final_Length3 <= 10 :
        font_size = 0.3
    elif final_Length3 > 10 and final_Length3 <= 20 :
        font_size = 0.5
    elif final_Length3 > 20 and final_Length3 <= 30 :
        font_size = 0.6
    elif final_Length3 > 30 :
        font_size = 0.8
           
    x_max_rect3= reference_point_x3 + final_Length3
    x_min_rect3= reference_point_x3 
    y_max_rect3= reference_point_y3 + final_Width3+3
    y_min_rect3= reference_point_y3
                
    d_x_L = x_min_rect3 - x_min_rect1
    d_x_R = x_max_rect1 - x_max_rect3
    d_y = y_max_rect1 - y_max_rect3

    Width3 = Width3 + d_y
  
    if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
        final_Length3 = Length3 + 2*H_f1
        final_Width3 = Width3 + 2*H_f1
        reference_point_x3 = x_min3 - H_f1
        reference_point_y3 = y_min3 - H_f1
                        
    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
        final_Length3 = Length3 + H_f1
        final_Width3 = Width3 + 2*H_f1
        reference_point_x3 = x_min3 - H_f1
        reference_point_y3 = y_min3 - H_f1
                        
    elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
        final_Length3 = Length3 + H_f1
        final_Width3 = Width3 + 2*H_f1
        reference_point_x3 = x_min3
        reference_point_y3 = y_min3 - H_f1
                       
    elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
        final_Length3 = Length3 + 2*H_f1
        final_Width3 = Width3 + H_f1
        reference_point_x3 = x_min3 - H_f1
        reference_point_y3 = y_min3
                        
    elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
        final_Length3 = Length3
        final_Width3 = Width3 + 2*H_f1
        reference_point_x3 = x_min3
        reference_point_y3 = y_min3 - H_f1
                       
    elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
        final_Length3 = Length3 + H_f1
        final_Width3 = Width3 + H_f1
        reference_point_x3 = x_min3 - H_f1
        reference_point_y3 = y_min3
                    
    elif free_side[1][0]== 0 and free_side[1][1]==1 and free_side[1][2]==1:
        final_Length3 = Length3 + H_f1
        final_Width3 = Width3 + H_f1
        reference_point_x3 = x_min3 
        reference_point_y3 = y_min3
                       
    else:
        final_Length3 = Length3
        final_Width3 = Width3 + H_f1
        reference_point_x3 = x_min3
        reference_point_y3 = y_min3
           
    origin = Vec3(reference_point_x3,reference_point_y3)
    rect_3 = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3,final_Width3+3),
                                    origin), close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
                
                
                
    PL_s_x1 = reference_point_x3 
    PL_s_y1 = reference_point_y3 + final_Width3
    PL_e_x1 = reference_point_x3 + final_Length3
    Pl_e_y1 = PL_s_y1
                
    PL_s_x2 = reference_point_x3
    PL_s_y2 = reference_point_y3 + final_Width3
    PL_e_x2 = reference_point_x3 + final_Length3
    PL_E_y2 = PL_s_y2
                
    PL_s_x3 = PL_s_x1
    PL_s_y3 = PL_s_y1
    PL_e_y3 = PL_s_y1 + 3
    PL_s_x4 =PL_e_x1
    PL_s_y4 = PL_s_y1
    PL_e_y4 = PL_s_y1 +3
                
            
    line_1_horizontal = target_models.add_line( (PL_s_x1 , PL_s_y1) ,(PL_e_x1 , PL_s_y1) , 
                                       dxfattribs = {'color': 7 , 'layer': 'nasb'} )
    line_3_horizontal = target_models.add_line( (PL_s_x2 , PL_s_y2) ,(PL_e_x2 , PL_s_y2) , 
                                       dxfattribs = {'color': free_side[1][3] , 'layer': 'nasb'} )
    line_1_vertical_L = target_models.add_line( (PL_s_x3 , PL_s_y3) ,(PL_s_x3 , PL_e_y3) , 
                                       dxfattribs = {'color': 7 , 'layer':'nasb'} )                
    line_1_vertical_R = target_models.add_line( (PL_s_x4 , PL_s_y4) ,(PL_s_x4 , PL_e_y4) , 
                                       dxfattribs = {'color': 7 , 'layer':'nasb'} )
    trans_rect3_x = - reference_point_x3
    trans_rect3_y = - reference_point_y3
                
    trans_nobject_x = - reference_point_x3
    trans_nobject_y = - reference_point_y3
                
    trans_line3horizontal_x = - reference_point_x3
    trans_line3horizontal_y = - reference_point_y3
    
    for enn in the_next_object:
        ennc = enn.copy()
        enn.translate(trans_nobject_x + x6 , trans_nobject_y+distance_y,0)
        enn.dxf.color = free_side[1][3]
        enn.dxf.layer = 'nasb'
        ennc.translate( trans_first_object_x  + x6 , trans_first_object_y + distance_y , 0 )
        ennc.dxf.color = 7
        ennc.dxf.layer = 'nasb'
        
        if free_side[1][4] == 1 :
            ss = ennc.copy()
            ss.translate(-final_Length1-5 , 0, 0 )
            ss.dxf.color = free_side[0][3]
            ss.dxf.layer = 'nasb'
            yy = ss.get_points()
            pnx = 0
            pny = 0 
            for ih in yy:
                if ih[0] > pnx and ih[1]> pny:
                    pnx = float(ih[0])
                    pny = float(ih[1])
            
            
            
            target_models.add_entity(ss)
            text = target_models.add_text(str(i+1), height = 0.4, 
                dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((pnx , pny) ,
                align = TextEntityAlignment.LEFT)
            #text.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
            
        target_models.add_entity(ennc)
        #print(pnx+1)
            
        
                
    rect_3.translate(trans_rect3_x + x6,trans_rect3_y+distance_y,0)
   
        
        
        
        
        
        
        #if free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==0:
        #    origin = Vec3(reference_point_x3+3 + H_i,reference_point_y3+3 + H_i)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3-6 - 2*H_i,
        #                final_Width3-6 - 2*H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
        #    x1s = reference_point_x3+3 + H_i
        #    x1e = reference_point_x3+3 + H_i
        #    y1s = reference_point_y3+3 + H_i 
        #    y1e = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    x2s = reference_point_x3+3 + H_i
        #    x2e = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    y2s = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    y2e = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    x3s = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    x3e = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    y3s = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    y3e = reference_point_y3+3 + H_i 
        
        #    line_1_vertical_Rd = target_models.add_line( (x1s,y1s) ,(x1e,y1e) , 
        #                               dxfattribs = {'color': 7 , 'layer': 'nasb'} )
        #    line_1_vertical_Rd.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)

        #    line_1_vertical_Ld = target_models.add_line( (x3s,y3s) ,(x3e,y3e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )    
        #    line_1_vertical_Ld.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+5,reference_point_y3+3 + H_i + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
            
                        
        #elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==0:
        #    origin = Vec3(reference_point_x3+3 + H_i,reference_point_y3+3 + H_i)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3-3 - H_i,
        #                final_Width3-6 - 2*H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
        #    x1s = reference_point_x3+3 + H_i
        #    x1e = reference_point_x3+3 + H_i
        #    y1s = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    y1e = reference_point_y3+3 + H_i 
        #    x2s = reference_point_x3+3 + H_i
        #    x2e = reference_point_x3+3 + H_i + final_Length3-3 - H_i
        #    y2s = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    y2e = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    x3s = reference_point_x3+3 + H_i
        #    x3e = reference_point_x3+3 + final_Length3-3
        #    y3s = reference_point_y3+3 + H_i
        #    y3e = reference_point_y3+3 + H_i 
        
        #    line_1_vertical_Rd = target_models.add_line( (x1s,y1s) ,(x1e,y1e) , 
        #                               dxfattribs = {'color': 7 , 'layer': 'nasb'} )
        #    line_1_vertical_Rd.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)

        #    line_1_horizental_Ld = target_models.add_line( (x3s,y3s) ,(x3e,y3e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )    
        #    line_1_horizental_Ld.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+5,reference_point_y3+3 + H_i + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
                        
        #elif free_side[1][0]==0 and free_side[1][1]==1 and free_side[1][2]==0:
        #    origin = Vec3(reference_point_x3,reference_point_y3+3 + H_i)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3-3 - H_i,
        #                final_Width3-6 - 2*H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
        #    x1s = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    x1e = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    y1s = reference_point_y3+3 + H_i 
        #    y1e = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
            
        #    x2s = reference_point_x3#+3 + H_i
        #    x2e = reference_point_x3 + final_Length3-3 - H_i
        #    y2s = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    y2e = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
            
        #    x3s = reference_point_x3#+3 + H_i
        #    x3e = reference_point_x3 + final_Length3 - 3 - H_i
        #    y3s = reference_point_y3+3 + H_i
        #    y3e = reference_point_y3+3 + H_i 
        #    y3e = reference_point_y3+3 + H_i 
        
        #    line_1_vertical_Rd = target_models.add_line( (x1s,y1s) ,(x1e,y1e) , 
        #                               dxfattribs = {'color': 7 , 'layer': 'nasb'} )
        #    line_1_vertical_Rd.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)

        #    line_1_horizental_Ld = target_models.add_line( (x3s,y3s) ,(x3e,y3e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )    
        #    line_1_horizental_Ld.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+5,reference_point_y3+3 + H_i + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
            
                       
        #elif free_side[1][0]==0 and free_side[1][1]==0 and free_side[1][2]==1:
        #    origin = Vec3(reference_point_x3+3 + H_i,reference_point_y3)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3-6 - 2*H_i,
        #                final_Width3-3 - H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
        #    x1s = reference_point_x3+3 + H_i
        #    x1e = reference_point_x3+3 + H_i
        #    y1s = reference_point_y3+ final_Width3 -3 - H_i - d_y 
        #    y1e = reference_point_y3
            
        #    x2s = reference_point_x3+3 + H_i
        #    x2e = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    y2s = reference_point_y3 + final_Width3 - 3 - H_i - d_y
        #    y2e = reference_point_y3 + final_Width3 - 3 - H_i - d_y
            
        #    x3s = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    x3e = reference_point_x3+3 + H_i + final_Length3-6 - 2*H_i
        #    y3s = reference_point_y3 + final_Width3 - 3 - H_i - d_y
        #    y3e = reference_point_y3 
        
        #    line_1_vertical_Rd = target_models.add_line( (x1s,y1s) ,(x1e,y1e) , 
        #                               dxfattribs = {'color': 7 , 'layer': 'nasb'} )
        #    line_1_vertical_Rd.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)

        #    line_1_vertical_Ld = target_models.add_line( (x3s,y3s) ,(x3e,y3e) , 
        #                              dxfattribs = {'color': 7 , 'layer':'nasb'} )    
        #    line_1_vertical_Ld.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+4,reference_point_y3+3 + H_i + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
    
                        
        #elif free_side[1][0]==1 and free_side[1][1]==1 and free_side[1][2]==0:
        #    origin = Vec3(reference_point_x3,reference_point_y3+3 + H_i)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3,
        #                final_Width3-6 - 2*H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
        #    x2s = reference_point_x3
        #    x2e = reference_point_x3+ final_Length3
        #    y2s = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    y2e = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+5,reference_point_y3+3 + H_i + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        
                       
        #elif free_side[1][0]==1 and free_side[1][1]==0 and free_side[1][2]==1:
        #    origin = Vec3(reference_point_x3+3 + H_i,reference_point_y3)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3-3 - H_i,
        #                final_Width3-3-H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
            
        #    x2s = reference_point_x3 + 3 + H_i
        #    x2e = reference_point_x3 + final_Length3
        #    y2s = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
        #    y2e = reference_point_y3+3 + H_i + final_Width3-6 - 2*H_i - d_y
            
        #    x3s = reference_point_x3 + 3 + H_i
        #    x3e = reference_point_x3 + 3 + H_i
        #    y3s = reference_point_y3 + final_Width3 - 3 - H_i - d_y
        #    y3e = reference_point_y3
        
            

        #    line_1_vertical_Ld = target_models.add_line( (x3s,y3s) ,(x3e,y3e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )    
        #    line_1_vertical_Ld.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
            
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+4,reference_point_y3+3 + H_i + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        
                    
        #elif free_side[1][0]== 0 and free_side[1][1]==1 and free_side[1][2]==1:
        #    origin = Vec3(reference_point_x3,reference_point_y3)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3-3 -H_i,
        #                final_Width3-3 - H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
        #    x1s = reference_point_x3 + final_Length3 - 3 - H_i
        #    x1e = reference_point_x3
        #    y1s = reference_point_y3 + final_Width3 - 3 - H_i - d_y
        #    y1e = reference_point_y3 + final_Width3 - 3 - H_i - d_y
            
        #    x2s = reference_point_x3 + final_Length3 - 3 - H_i
        #    x2e = reference_point_x3 + final_Length3 - 3 - H_i
        #    y2s = reference_point_y3 + final_Width3-3 - H_i - d_y
        #    y2e = reference_point_y3
            
           
        
        #    line_1_vertical_Rd = target_models.add_line( (x1s,y1s) ,(x1e,y1e) , 
        #                               dxfattribs = {'color': 7 , 'layer': 'nasb'} )
        #    line_1_vertical_Rd.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)

           
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+5,reference_point_y3 + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y+final_Width3 - 5 - H_i - d_y , 0)
        
                       
        #else:
        #    origin = Vec3(reference_point_x3,reference_point_y3)
        #    rect_3dola = target_models.add_lwpolyline(forms.translate(forms.box(final_Length3,
        #                final_Width3-3 - H_i - d_y),origin), 
        #                close = True , dxfattribs= {'color': free_side[1][3] , 'layer': 'nasb'})
        #    rect_3dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    dola.dxf.color = free_side[1][3]
        #    target_models.add_entity(dola)
        
            
        #    x2s = reference_point_x3
        #    x2e = reference_point_x3 + final_Length3
        #    y2s = reference_point_y3 + final_Width3 - 3 - H_i - d_y
        #    y2e = reference_point_y3+ final_Width3 - 3 - H_i - d_y
            
           
        #    line_1_horizontal_d = target_models.add_line( (x2s,y2s) ,(x2e,y2e) , 
        #                               dxfattribs = {'color': 7 , 'layer':'nasb'} )
        #    line_1_horizontal_d.translate(trans_rect3_x + x6 , trans_rect3_y+distance_y , 0)
        #    text = target_models.add_text(str(i+1), height = 1, 
        #        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((
        #        reference_point_x3+5,reference_point_y3 + 0.5) ,
        #        align = TextEntityAlignment.LEFT)
        #    text.translate(-(Length3+10)+trans_rect3_x + x6 , trans_rect3_y
        #                   +distance_y+final_Width3 - 5 - H_i - d_y , 0)
        
        
        
        
            
                
    line_3_horizontal_translated = line_3_horizontal.translate(trans_line3horizontal_x + x6,
    trans_line3horizontal_y+distance_y,0)
    line1verticalr = line_1_vertical_R.translate(trans_rect1_x + x6,trans_rect1_y+distance_y , 0)
    line1verticall = line_1_vertical_L.translate(trans_rect1_x + x6,trans_rect1_y+distance_y,0)
    line1horizontal = line_1_horizontal.translate(trans_rect1_x + x6,trans_rect1_y+distance_y,0)
      

    
    target_models.add_text(str(substrate_name), height = font_size , 
            dxfattribs = {'layer':'nasb', 'style':'TXT','color':4}).set_placement((final_Length3/2 + x6,distance_y+2) ,
            align = TextEntityAlignment.CENTER)
    target_models.add_text('code: '+str(code) , height = font_size ,
            dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((final_Length3/2 + x6,distance_y+1) ,
            align = TextEntityAlignment.CENTER)
    target_models.add_text(str(i+1), height = 1, 
                        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((1 + x6,distance_y+1) ,
                        align = TextEntityAlignment.LEFT)
    
    target_models.add_text(str(i+1), height = 1, 
                        dxfattribs ={'layer':'nasb','style':'TXT','color':4}).set_placement((1 + x6,
                        final_Width3 + distance_y+1) , align = TextEntityAlignment.LEFT)
    target_models.add_text(str(i+1), height = 1, 
                        dxfattribs = {'layer':'nasb','style':'TXT','color':4}).set_placement((final_Length1 + 6 + x6 ,
                        distance_y+1) ,align = TextEntityAlignment.LEFT)
    target_models.add_text(str(i+1), height = 1, 
                        dxfattribs= {'layer':'nasb','style':'TXT','color':4}).set_placement((2*final_Length1 + 11 + x6,
                        distance_y+1) ,align = TextEntityAlignment.LEFT)
  
                
    target_doc.saveas(tar_path)
    
    file_path = os.path.expanduser("C:/Users/A.Mohammadlou/Desktop/software_test/test.dxf")
    os.remove(file_path)
    update_i_distance_nasb(distance_y ,i ,final_Width1 , substrate_name , x6)
    

def find_the_last_file_cnc (x1 , x2 , x3 , x4 , x5 , x6):
    folder_path = "C:/Users/A.Mohammadlou/Desktop/"+str(the_code)+'/اندازه تولید/'
    files_a = os.listdir(folder_path)
    files = [f for f in files_a if f.endswith(".dxf")]
    if files:
        files.sort(key=lambda x: os.path.getmtime(os.path.join(folder_path, x)))
        last_file = files[-1]
    last_file_path = folder_path + str(last_file)
    create_the_cnc_sub(x1 , x2 , x3 , x4 , last_file_path , x5 , x6)

def find_the_last_file_nasb (x1 , x2 , x3 , x4 , x5 , x6):
    folder_path = "C:/Users/A.Mohammadlou/Desktop/"+str(the_code)+'/اندازه تولید/'
    files_a = os.listdir(folder_path)
    files = [f for f in files_a if f.endswith(".dxf")]
    if files:
        files.sort(key=lambda x: os.path.getmtime(os.path.join(folder_path, x)))
        last_file = files[-1]
    last_file_path = folder_path + str(last_file)
    create_the_nasb_sub(x1 , x2 , x3 , x4 , last_file_path , x5 , x6)

def take_i_distance_cnc(x , y):
    with open('C:/Users/A.Mohammadlou/Desktop/software_test/previous_values.txt', 'r') as f:
        contents = f.read()
        numbers_str = contents.split('\n')
            
        i_value = int(numbers_str[0])
        distance_c = float(numbers_str[2])
        distance_x = float(numbers_str[4])
    f.close()
    find_the_last_file_cnc(distance_c , i_value , x , the_code , y , distance_x)

    
def take_i_distance_nasb(x , y):
    with open('C:/Users/A.Mohammadlou/Desktop/software_test/previous_values.txt', 'r') as f:
        contents = f.read()
        numbers_str = contents.split('\n')
            
        i_value = int(numbers_str[0])
        distance_y = float(numbers_str[1])
        distance_x = float(numbers_str[3])
    f.close()
    find_the_last_file_nasb(distance_y , i_value , x , the_code , y , distance_x)
    
    
    
    
    


def get_name_of_sub_cnc():
    sub_name = entry3.get()
    f_c = s_m_f.get()
    n_c = s_m_n.get()
    f_s_fn = []
    f_s_f = []
    f_s_n = []
    
    f_r = ch_vf_right.get()

    f_s_f.append(f_r)
    f_l = ch_vf_left.get()
    f_s_f.append(f_l)
    f_d = ch_vf_down.get()
    f_s_f.append(f_d)
    n_r = ch_vn_right.get()
    f_s_n.append(n_r)
    n_l = ch_vn_left.get()
    f_s_n.append(n_l)
    n_d = ch_vn_down.get()
    f_s_n.append(n_d)
    
    if f_c == "Karji" or f_c == "MDF":
        f_c_i = 1
    elif f_c == "Texon 0.5":
        f_c_i = 3
    elif f_c == "Texon 1":
        f_c_i = 84
    else:
        f_c_i = 2
    f_s_f.append(f_c_i)
    f_do = ch_vf_dola.get()
    f_s_f.append(f_do)
        
        
    if n_c == "Karji" or n_c == "MDF":
        n_c_i = 1
    elif n_c == "Texon 0.5":
        n_c_i = 3
    elif n_c == "Texon 1":
        n_c_i = 84
    else:
        n_c_i = 2
    f_s_n.append(n_c_i)
    n_do = ch_vn_dola.get()
    f_s_n.append(n_do)
    n_yg = ch_vn_yg.get()
    f_s_n.append(n_yg)
    n_mg = ch_vn_mg.get()
    f_s_n.append(n_mg)
    n_si = ch_vn_si.get()
    f_s_n.append(n_si)
    n_nm = ch_vn_nm.get()
    f_s_n.append(n_nm)
    n_rt = entry_rt.get() #9 rotate
    f_s_n.append(n_rt)
    n_mr = entry_mr.get() #10 mirror
    f_s_n.append(n_mr)
        
    
    f_s_fn.append(f_s_f)
    f_s_fn.append(f_s_n)
    
    if sub_name == '':
        window4 = tkinter.Tk()
        window4.withdraw()
        messagebox.showerror("Name Error", "Enter a name !")
    elif the_dict.get(sub_name) is None:
        take_i_distance_cnc(sub_name , f_s_fn)
    else:
        window5 = tkinter.Tk()
        window5.withdraw()
        messagebox.showerror("Name Error", "This name is already exist !")

def get_name_of_sub_nasb():
    sub_name = entry3.get()
    f_c = s_m_f.get()
    n_c = s_m_n.get()
    f_s_fn = []
    f_s_f = []
    f_s_n = []
    
    f_r = ch_vf_right.get()

    f_s_f.append(f_r)
    f_l = ch_vf_left.get()
    f_s_f.append(f_l)
    f_d = ch_vf_down.get()
    f_s_f.append(f_d)
    n_r = ch_vn_right.get()
    f_s_n.append(n_r)
    n_l = ch_vn_left.get()
    f_s_n.append(n_l)
    n_d = ch_vn_down.get()
    f_s_n.append(n_d)
    
    
    if f_c == "Karji" or f_c == "MDF":
        f_c_i = 1
    elif f_c == "Texon 0.5":
        f_c_i = 3
    elif f_c == "Texon 1":
        f_c_i = 84
    else:
        f_c_i = 2
    f_s_f.append(f_c_i)
    f_do = ch_vf_dola.get()
    f_s_f.append(f_do)
        
        
    if n_c == "Plexi":
        n_c_i = 2
    elif n_c == "Texon 0.5":
        n_c_i = 3
    elif n_c == "Texon 1":
        n_c_i = 84
    else:
        n_c_i = 1
    f_s_n.append(n_c_i)
    n_do = ch_vn_dola.get()
    f_s_n.append(n_do)
    n_rt = entry_rt.get()
    f_s_n.append(n_rt) #5 rotate
    n_mr = entry_mr.get()
    f_s_n.append(n_mr) #6 mirror
        
    
    f_s_fn.append(f_s_f)
    f_s_fn.append(f_s_n)
    
    if sub_name == '':
        window4 = tkinter.Tk()
        window4.withdraw()
        messagebox.showerror("Name Error", "Enter a name !")
        
    elif the_dict.get(sub_name) is None:
        take_i_distance_nasb(sub_name , f_s_fn)
    else:
        window4 = tkinter.Tk()
        window4.withdraw()
        messagebox.showerror("Name Error", "This name is already exist !")
        
    
    
    
    
      
def the_name_of_sub():
    global ch_vf_right
    global ch_vf_left
    global ch_vf_down
    global ch_vn_right
    global ch_vn_left
    global ch_vn_down
    global ch_vf_dola
    global ch_vn_dola
    global ch_vn_yg
    global ch_vn_mg
    global ch_vn_si
    global ch_vn_nm
    global entry3
    global entry_rt
    global entry_mr
    global s_m_f
    global s_m_n
    window2 = tkinter.Tk()
    window2.geometry('320x400')
    window2.title('Nasb & Cnc')
    frame2 = tkinter.Frame(window2)
    frame2.pack(padx=1, pady=1)
    
    label3_0 = tkinter.Label(frame2 , text= 'Select the Side to be free :')
    label3_0.grid(row=0, column=0,columnspan=2, padx=1, pady=1)
    
    label3_1 = tkinter.Label(frame2 , text= 'First object')
    label3_1.grid(row=1, column=0, padx=1, pady=1)
    label3_2 = tkinter.Label(frame2 , text= 'Next object')
    label3_2.grid(row=1, column=1, padx=1, pady=1)

    ch_vf_right = tkinter.IntVar(frame2)
    ch_vn_right = tkinter.IntVar(frame2)
    
    ch_bf_right =tkinter.Checkbutton(frame2, text= 'Right' , variable=ch_vf_right , width=3)
    ch_bn_right =tkinter.Checkbutton(frame2, text= 'Right ' , variable=ch_vn_right , width=3)
    
    ch_bf_right.grid(row=2, column=0, padx=1, pady=1)
    ch_bn_right.grid(row=2, column=1, padx=1, pady=1)
    

    
    ch_vf_left = tkinter.IntVar(frame2)
    ch_vn_left = tkinter.IntVar(frame2)
    
    ch_bf_left =tkinter.Checkbutton(frame2, text= 'Left  ' , variable=ch_vf_left , width=4)
    ch_bn_left =tkinter.Checkbutton(frame2, text= 'Left  ' , variable=ch_vn_left , width=3)
    
    ch_bf_left.grid(row=3, column=0, padx=1, pady=1)
    ch_bn_left.grid(row=3, column=1, padx=3, pady=1)

    

    
    
    ch_vf_down = tkinter.IntVar(frame2)
    ch_vn_down = tkinter.IntVar(frame2)
    
    ch_bf_down =tkinter.Checkbutton(frame2, text= 'Down' , variable=ch_vf_down , width=3)
    ch_bn_down =tkinter.Checkbutton(frame2, text= 'Down' , variable=ch_vn_down , width=3)
    
    ch_bf_down.grid(row=4, column=0, padx=1, pady=1)
    ch_bn_down.grid(row=4, column=1, padx=4, pady=1)
    
    ch_vf_dola = tkinter.IntVar(frame2)
    ch_vn_dola = tkinter.IntVar(frame2)
    
    ch_bf_dola =tkinter.Checkbutton(frame2, text= 'Dola' , variable=ch_vf_dola , width=3)
    ch_bn_dola =tkinter.Checkbutton(frame2, text= 'Dola' , variable=ch_vn_dola , width=3)
    
    ch_bf_dola.grid(row=5, column=0, padx=1, pady=1)
    ch_bn_dola.grid(row=5, column=1, padx=4, pady=1)
    
    label3_3 = tkinter.Label(frame2 , text= '------------------------------')
    label3_3.grid(row=6, column=0,columnspan=2, padx=1, pady=1)
    
    ch_vn_yg = tkinter.IntVar(frame2)
    ch_vn_mg = tkinter.IntVar(frame2)
    
    
    
    ch_bn_yg =tkinter.Checkbutton(frame2, text= 'Yaragh' , variable=ch_vn_yg , width=4)
    ch_bn_yg.grid(row=7, column=1, padx=4, pady=1)
    ch_bn_mg =tkinter.Checkbutton(frame2, text= 'Maghzi' , variable=ch_vn_mg , width=4)
    ch_bn_mg.grid(row=7, column=0, padx=4, pady=1)
    
    ch_vn_si = tkinter.IntVar(frame2)
    ch_vn_nm = tkinter.IntVar(frame2)
    
    ch_bn_si =tkinter.Checkbutton(frame2, text= 'Didd' , variable=ch_vn_si , width=4)
    ch_bn_si.grid(row=8, column=0, padx=4, pady=1)
    ch_bn_nm =tkinter.Checkbutton(frame2, text= 'Nama' , variable=ch_vn_nm , width=4)
    ch_bn_nm.grid(row=8, column=1, padx=4, pady=1)
    
    #ch_vn_si = tkinter.IntVar(frame2)
    #ch_vn_nm = tkinter.IntVar(frame2)
    
    label4 = tkinter.Label(frame2, text='Rotate(angle):')
    label4.grid(row=9, column=0, padx=1, pady=1)
    
    label5 = tkinter.Label(frame2, text='Mirror(x/y):' , width=10)
    label5.grid(row=9, column=1, padx=1, pady=1)
    
    entry_rt = tkinter.Entry(frame2 ,  width=5)
    entry_rt.grid(row=10, column=0, padx=1, pady=1)
    
    entry_mr = tkinter.Entry(frame2 ,  width=5)
    entry_mr.grid(row=10, column=1, padx=1, pady=1)
    #entry_rt.place(x=20, y=250)
    
    
    label3_3 = tkinter.Label(frame2 , text= 'Select Materials Type')
    label3_3.grid(row=11, column=0,columnspan=2, padx=1, pady=1)
    
    s_m_f = tkinter.StringVar(frame2)
    s_m_n = tkinter.StringVar(frame2)
    
    combo_f = ttk.Combobox(frame2, values=["Karji", "Texon 0.5", "Texon 1" , "Plexi" , "MDF"],
                          textvariable=s_m_f , width=8)
    combo_n = ttk.Combobox(frame2, values=["Karji", "Texon 0.5", "Texon 1" , "Plexi" , "MDF"], 
                          textvariable=s_m_n, width=8)
    combo_f.grid(row=12 , column=0 , padx=1 , pady=1)
    combo_n.grid(row=12 , column=1 , padx=1 , pady=1)
    
    
    
    label3 = tkinter.Label(frame2, text='Spicify a name for your substrate')
    label3.grid(row=13, column=0, columnspan=2, padx=1, pady=1)
    entry3 = tkinter.Entry(frame2)
    entry3.grid(row=14, column=0, columnspan=2, padx=1, pady=1)
    
    
    
    button2 = tkinter.Button(frame2, text='nasb' , command =get_name_of_sub_nasb)
    button2.grid(row=15, column=0, padx=1, pady=1)
    button3 = tkinter.Button(frame2, text='cnc' , command =get_name_of_sub_cnc)
    button3.grid(row=15, column=1, padx=1, pady=1)
 
    



def create_the_document(x1):
    global tar_path
    source_path = 'C:/Users/A.Mohammadlou/Desktop/information bank'
    destination_path = 'C:/Users/A.Mohammadlou/Desktop/'+str(the_code)+'/مقر/'

    if not os.path.exists(destination_path):
        os.makedirs(destination_path)
    
    cc = 0
    for filename in os.listdir(destination_path):
        if filename == str(x1)+'.dxf':
            tar_path = os.path.join(destination_path, filename)
            target_doc = ezdxf.readfile(tar_path)
            target_models = target_doc.modelspace()
            if target_doc.styles.has_entry(name='TXT') == False:
                target_doc.styles.new('TXT', dxfattribs = {'font':'txt.shx'})
            else:
                pass
            points_list = []
            strings = []
            numbers = []
            for eenn in target_models:
                if eenn.dxftype() == 'LWPOLYLINE':
                    points_list += eenn.get_points()
                elif eenn.dxftype() == 'TEXT':
                    strings.append(eenn.plain_text())
                    
            for string in strings:
                if string.isnumeric() and len(string) <= 2:
                    numbers.append(int(string))

            if len(numbers) > 0 :
                max_number = max(numbers)
            else:
                max_number = 0
            distance_x = 0.0
            distance_y = 0.0
            distance_c = 0.0
            i_value = max_number

            for point in points_list:
                if point[0] > distance_x:
                    distance_x = point[0]
            if distance_x > 0 :
                distance_x += 10
            else:
                pass
            cc = 1
            distance_x_c = distance_x + 600
        else:
            pass
        
    if cc == 0 :
        shutil.copy(os.path.join(source_path, 'newfile.dxf' ), os.path.join(destination_path, str(x1)+'.dxf'))
        tar_path = destination_path + str(x1) + '.dxf'
        i_value = 0
        distance_c = 0
        distance_x = 0
        distance_y = 0
        distance_x_c = 600
    else:
        pass
    
    
   
    with open('C:/Users/A.Mohammadlou/Desktop/software_test/previous_values.txt', 'w') as f:
        f.write(str(i_value))
        f.write('\n')
        f.write(str(distance_y))
        f.write('\n')
        f.write(str(distance_c))
        f.write('\n')
        f.write(str(distance_x))
        f.write('\n')
        f.write(str(distance_x_c))
    f.close()

    
    the_name_of_sub()
    
    
    
    
 
  


def get_entry():
    global the_code
    name_of_document = entry1.get()
    the_code = entry2.get()
    create_the_document(name_of_document)
    
    
    
    
    
window1 = tkinter.Tk()
window1.geometry('400x300')
window1.title('Aleekhan ---  bassamrabihavi@gmail.com')
frame1 = tkinter.Frame(window1)
frame1.pack()
label1 = tkinter.Label(frame1 , text='Spicify the name of the new dxf file ')
label1.pack()
entry1 = tkinter.Entry(frame1)
entry1.pack()

label2 = tkinter.Label(frame1 , text= 'Enter the code of your product ')
label2.pack()
entry2 = tkinter.Entry(frame1)
entry2.pack()



tree = ttk.Treeview(window1, columns=('0', '1'), selectmode='browse')
tree.heading('#0', text='name')
tree.heading('#1', text='number')
tree.column('#0', width=300)
tree.column('#1', width=50)
tree.pack(padx=10, pady=10)



button1 = tkinter.Button(frame1, text="OK", command=get_entry)
button1.pack()


window1.mainloop()
```
